## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from itertools import permutations
```

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

L = list(sorted(list(permutations("0123456789")))[10**6 -1])

answer = "".join(L)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Time Complexity:
   - Generating Permutations: The `permutations` function generates all permutations of the characters in the string $"0123456789"$. The time complexity of this function is $O(n!)$, where $n$ is the length of the string. Generating all permutations has an exponential time complexity.
   - Sorting Permutations: The `sorted` function is used to sort the generated permutations. Sorting the permutations has a time complexity of $O(n! *log(n!))$, which is also exponential.
   - Accessing the Element: The code accesses the element at the index ($10^6 - 1$) from the sorted list of permutations. This operation has a constant time complexity of $O(1)$.

   Therefore, the overall time complexity of the code is dominated by the generation and sorting of permutations, resulting in an approximate time complexity of $O(n! *log(n!))$.

2. Space Complexity:
   - Permutations List: The code generates all permutations of the characters in the string $"0123456789"$ and stores them in a list. The space complexity of this list is $O(n!)$, where n is the length of the string. Storing all permutations requires significant memory as the number of permutations grows exponentially.
   - Sorted List: The code creates a sorted list of permutations. The space complexity of this list is also $O(n!)$, as it contains all the generated permutations.

   Overall, the space complexity of the code is $O(n!)$ due to the storage of the permutations and the sorted list.
