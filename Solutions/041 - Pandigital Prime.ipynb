## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from itertools import permutations  #generate permutations
```

## Main Solution <a name="Main-Solution"></a>
First we take care of the prime part of the problem with below definition:

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+ 1):
        if n%i == 0:
            return False
    return True
```

Then we try generate permutations from the largest possible numbers and simultaneously check prime condition.

```python
start_time = time.time()

digits = [9,8,7,6,5,4,3,2,1]
maximum = 0
while len(digits)!=4:
    for n in permutations(digits):
        if n[0] == digits[0]:
            number = int("".join(str(i) for i in n))
            if is_prime(number):
                maximum = max(maximum,number)

    del digits[0]

answer = maximum

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n!)$**, where $n$ is the number of digits in the list. This is because the code uses the permutations function from the itertools module, which generates all possible permutations of the digits. The number of permutations is factorial of the number of digits, so the time complexity is $O(n!)$.

The **space complexity** of my solution is **$O(n)$**, where $n$ is the number of digits in the list. This is because the code creates a list of digits and a variable to store the maximum prime number found. The space used by these variables is proportional to the number of digits, so the space complexity is $O(n)$.
