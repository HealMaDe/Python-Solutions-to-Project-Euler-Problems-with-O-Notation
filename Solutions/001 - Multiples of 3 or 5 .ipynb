## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

L = []
for i in range(1,1000):
    if i%3 == 0 or i%5 ==0:
        L.append(i)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:", sum(L),"\nExecution time:", "{:.5f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution for this problem is **$O(n)$**, where $n$ is the range of numbers from $1$ to $1000$. This is because the code iterates through each number in the range once and performs a constant amount of operations (checking if the number is divisible by $3$ or $5$ and appending it to the list if it is).

The **space complexity** of my solution for this problem is also **$O(n)$**, as the size of the list $L$ will grow linearly with the number of elements that meet the condition (divisible by $3$ or $5$). In the worst case scenario, all numbers in the range could meet the condition and be added to the list.
