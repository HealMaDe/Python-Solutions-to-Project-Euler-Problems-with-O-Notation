## Main Solution <a name="Main-Solution"></a>
First we difine our functions. Remeber we have to define memoization function to use it as a look up table for calculated Collatz sequence

```python
def collatz(n):
    if n%2 == 0:
        return n//2
    else:
        return 3*n + 1

def memoization(n,memory):
    if n not in memory:
        memory[n] = memoization(collatz(n),memory) + 1
    return memory[n]
```

Now we try to solve the problem

```python
start_time = time.time()

memory = {1:1} # A memory that contain first term of the collatz sequence with its chain
max = 0
for n in range(1,10**6):
    chain = memoization(n,memory)
    if chain > max:
        max = chain
        answer = n

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time complexity:
- The `collatz()` function performs a simple arithmetic calculation and a modulo operation, which can be considered as constant time operations.
- The `memoization()` function recursively calls itself until it finds a value in the `memory` dictionary. The number of recursive calls depends on the length of the Collatz sequence for the given input `n`.
- In the worst case, when `n` is a large number, the length of the Collatz sequence can be long. However, due to the memoization technique, once a value is computed, it is stored in the `memory` dictionary and reused. Therefore, the `memoization()` function will not recompute the same value multiple times.
- The overall time complexity of the code is determined by the number of unique values in the range from $1$ to $10^6$ that need to be computed and stored in the `memory` dictionary.
- In the worst case, all unique values from 1 to 10^6 need to be computed, resulting in $O(10^6)$ or simply $O(n)$ time complexity.

Space complexity:
- The code uses a dictionary `memory` to store the computed values of the Collatz sequence lengths. The number of entries in the dictionary corresponds to the unique values of `n` encountered during the execution of the code.
- In the worst case, when all unique values from $1$ to $10^6$ need to be stored, the `memory` dictionary will have approximately $10^6$ entries.
- Therefore, the space complexity of the `memory` dictionary is $O(n)$.

- The code also uses a few integer variables (`n`, `max`, and `answer`), which require constant space.

- The space complexity of the code is dominated by the `memory` dictionary, resulting in $O(n)$ space complexity.
