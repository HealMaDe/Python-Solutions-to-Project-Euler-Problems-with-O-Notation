## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

file = open("0022_names.txt","r")

# Read the contents of the file
content = file.read()
names = [i for i in content.replace('"', '').split(",")]

# Sort the names in alphabetical order
names.sort()

alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# dictionary to calculate the alphabetical value of a name
d = {}
for i in range(len(alphabet)):
    d[alphabet[i]]= i + 1

#calculating scores
score = 0
for index in range(len(names)):
    sum = 0
    for letter in names[index]:
        sum += d[letter]

    score += sum*(index + 1)

answer = score

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the number of names in the file. This is because there is a nested loop where the outer loop iterates over each name and the inner loop iterates over each letter in the name. Therefore, for each name, we iterate over all the letters in that name.


The **space complexity** of my solution is **$O(n)$**, where $n$ is the number of names in the file. This is because we store the names in a list, which requires space proportional to the number of names. Additionally, we store the alphabetical values of each letter in a dictionary, which also requires space proportional to the number of names.
