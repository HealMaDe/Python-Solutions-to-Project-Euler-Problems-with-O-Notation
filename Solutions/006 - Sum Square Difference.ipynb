## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

def f(n):  #sum of square
    L1 = []
    L2 = []
    for i in range(1,n+1):
        L2.append(i)
        L1.append(i**2)

    return sum(L2)**2 - sum(L1)

answer = f(100)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the input parameter. This is because the function uses a for loop that iterates $n$ times. Within each iteration, the function performs constant time operations such as appending to lists and calculating the square of a number.

The **space complexity** of my solution is **$O(n)$**, where $n$ is the input parameter. This is because the function creates two lists, $L1$ and $L2$, each of which can potentially contain $n$ elements. Therefore, the space required by the function grows linearly with the input size.
