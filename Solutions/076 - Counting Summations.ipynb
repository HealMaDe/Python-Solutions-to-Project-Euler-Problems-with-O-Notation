## Main Solution <a name="Main-Solution"></a>

All we have to do is to define a time-efficient function that counts number of ways to partition any given number $n$

```python
start_time = time.time()

def count_partition_ways(n):
    # Initialize the partitions array
    partitions = [0] * (n + 1)
    partitions[0] = 1

    # Calculate the partition function values
    for i in range(1, n):
        for j in range(i, n + 1):
            partitions[j] += partitions[j - i]

    return partitions[-1]

answer = count_partition_ways(100)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code uses a nested loop to calculate the partition function values.
- The outer loop iterates from 1 to `n - 1`, which has a complexity of $O(n)$.
- The inner loop iterates from `i` to `n + 1`, which has a complexity of $O(n)$.
- Therefore, the time complexity of the code is $O(n^2)$.

Space Complexity:
- The code initializes a list `partitions` of size `n + 1` to store the partition function values.
- The space required for the list is directly proportional to `n`.
- Therefore, the space complexity of the code is $O(n)$.
