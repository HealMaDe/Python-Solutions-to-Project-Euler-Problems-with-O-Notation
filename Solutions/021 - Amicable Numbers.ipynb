## Main Solution <a name="Main-Solution"></a>
First we take care of "the sum of proper divisors" with below definition:

```python
def sum_divisors(n):
    sum = 0
    for divisor in range(1,n//2 + 1):
        if n % divisor == 0:
            sum += divisor
    return sum
```

Now we can solve the problem.

```python
start_time = time.time()

answer = 0
for a in range(1,10000):
    b = sum_divisors(a)
    if sum_divisors(b) == a and a != b: #Amicable Condition
        answer += a

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the input number. This is because the outer loop iterates from 1 to $n$, and for each iteration, the inner loop iterates from $1$ to $n/2$. Therefore, the total number of iterations is approximately $n * (n/2) = n^2/2$.

The **space complexity** of my solution is **$O(1)$**, as it only uses a constant amount of extra space to store the variables $answer$, $sum$, $a$, and $b%. The space used does not depend on the size of the input number.
