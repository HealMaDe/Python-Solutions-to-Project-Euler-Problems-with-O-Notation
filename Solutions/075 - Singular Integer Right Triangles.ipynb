## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

limit = 15*10**5

# Initialize count for each possible L
count = [0] * (limit + 1)

# Generate Pythagorean triples using Euclid's formula
# a = m^2 - n^2, b = 2mn, c = m^2 + n^2
for m in range(2, int((limit // 2) ** 0.5) + 1):
    for n in range(1, m):
        if (m - n) % 2 == 1 == gcd(m, n) :
            a = m * m - n * n
            b = 2 * m * n
            c = m * m + n * n
            p = a + b + c
            while p <= limit:
                count[p] += 1
                p += a + b + c

# Count the number of L values with exactly one solution
answer = count.count(1)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The outer loop iterates from $2$ to the square root of $L/2$, which has a complexity of $O(\sqrt{L})$.
- The inner loop iterates from $1$ to $m$, which has a complexity of $O(m)$.
- The nested loops generate Pythagorean triplets using Euclid's formula, and the number of generated triplets is bounded by the square root of $L$.
- Therefore, the overall time complexity of the nested loops is approximately $O(\sqrt{L} * \sqrt{L}) = O(L)$.

- The code then iterates over the `count` list, which has a length of `limit + 1`. The time complexity of this iteration is $O(limit)$.

- The `count.count(1)` operation counts the number of occurrences of the value 1 in the `count` list. The time complexity of this operation is $O(limit)$.

- The overall **time complexity** of the code is dominated by the nested loops, resulting in a time complexity of $O(L)$.

Space Complexity:
- The code initializes a list `count` with a length of `limit + 1` to store the count of values for each possible $L$.
- The space required for the list is proportional to the value of `limit`.
- Therefore, the space complexity of the list is $O(L)$.

- The code also initializes some variables (`m`, `n`, `a`, `b`, `c`, `p`) that require constant space.

In our problem, the value of `limit` is $15*10^5$, so the time complexity of the code is $O(L)$, and the space complexity is also $O(L)$.

Note that We assume that the `gcd` function and the `count` function have constant time complexity.
