**$Hint$**

To find the largest prime factor of a given number, we can use a prime factorization algorithm. The main idea is to effectively divides the number by its prime factors, starting from the smallest prime $(2)$ and moving to larger primes. By dividing the number by its factors, we gradually reduce it until we are left with only prime factors.

This approach relies on the fact that any composite number can be expressed as a product of prime factors. By iterating through potential divisors and dividing the number, we can find all the prime factors that make up the original number.

For further reading about this Algorithm, check [here](https://geeksforgeeks.org/print-all-prime-factors-of-a-given-number/).

## Main Solution <a name="Main-Solution"></a>

Steps of the algorithm:

1.   Start with the given number.
2.   Divide the number by $2$ repeatedly until it is no longer divisible by $2$. Keep track of the number of divisions by $2$.
3.   Check divisibility by odd numbers starting from $3$ up to the square root of the number. If a divisor is found, divide the number by that divisor and add it to the list of prime factors.
4.   Continue checking divisibility by increasing odd numbers until the divisor exceeds the square root of the number.
5.   If the remaining number is greater than $2$, it is a prime factor, so add it to the list of prime factors.
6.   The resulting list contains all the prime factors of the given number.

<br>

Largest Prime Factor (LPF)  definition based on the above instructions

```python
def LPF(number):
    # Divide by 2 until it is no longer divisible
    while number % 2 == 0:
        Lpf = 2      #Larfest prime factor
        number //= 2

    # Check divisibility by odd numbers up to square root of the number
    for i in range(3, int(number**0.5) + 1, 2):
        while number % i == 0:
            Lpf = i
            number //= i

    # If the number itself is prime, it is the largest prime factor
    if number > 2:
        Lpf = number

    return Lpf
```

Finding The Answer

```python
start_time = time.time()

answer = LPF(600851475143)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(\sqrt{n})$**, where $n$ is the
input number. This is because the function checks divisibility by odd numbers up to the square root of the number, which takes $O(\sqrt{n})$ iterations.

The **space complexity** of my solution is **$O(1)$**, as it only uses a constant amount of additional space to store the largest prime factor.
