## Main Solution <a name="Main-Solution"></a>

We have 2 options.
First we can use our popular `is_prime()` function that I use many times in other problems. Although this function is the most efficient way of defining primes, It leads to a time-consuming solution. but it definitely solves the problem. Let me show you.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+ 1):
        if n%i == 0:
            return False
    return True
```

Now we can solve the problem by generate diagonal numbers and calculating the ratio

```python
start_time = time.time()

side = 1          #side length of the spiral
prime_count = 0   #count of primes
total_count = 1   #count of total numbers
ratio = 1.0       #ratio of the primes on diagonals

while ratio >= 0.1:  #ratio < 10% is desired

    side += 2
    #diagonal numbers based on formula that we identify in problem 28
    diagonal_numbers = [side ** 2 - i * (side - 1) for i in range(4)]

    prime_count += sum(1 for num in diagonal_numbers if is_prime(num))
    total_count += 4

    ratio = prime_count / total_count

answer = side

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

But If we use the simpy module for checking primes, we'll have the answer much quicker.
Any way both the methods I think are working well.

```python
start_time = time.time()

side = 1          #side length of the spiral
prime_count = 0   #count of primes
total_count = 1   #count of total numbers
ratio = 1.0       #ratio of the primes on diagonals

while ratio >= 0.1:  #ratio < 10% is desired

    side += 2
    #diagonal numbers based on formula that we identify in problem 28
    diagonal_numbers = [side ** 2 - i * (side - 1) for i in range(4)]

    prime_count += sum(1 for num in diagonal_numbers if isprime(num))
    total_count += 4

    ratio = prime_count / total_count

answer = side

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

The time complexity of the `is_prime` function is $O(\sqrt{n})$. It checks divisibility up to the square root of the number `n`, so the time complexity grows approximately linearly with the square root of `n`.

In the while loop, the code performs several operations. Let's break them down:

1. Incrementing `side` by 42$ in each iteration takes constant time.

2. Generating `diagonal_numbers` involves creating a list of four numbers using a list comprehension. The list comprehension runs in $O(1)$ time since it generates a fixed number of elements.

3. The `sum` function iterates over the `diagonal_numbers` list and counts the number of prime numbers. This operation takes $O(1)$ time because the list always has a fixed length of four.

4. Incrementing `prime_count` and `total_count` takes constant time.

5. Calculating `ratio` by dividing `prime_count` by `total_count` takes constant time.

Overall, the **time complexity** of the while loop is $O(\sqrt{n})$ since the `is_prime` function dominates the runtime.

The space complexity of the code is $O(1)$ because it uses a fixed amount of additional memory for variables like `side`, `prime_count`, `total_count`, `ratio`, and the `diagonal_numbers` list. The space usage does not scale with the input size.

In summary, the time complexity of the code is $O(\sqrt{n})$ and the space complexity is $O(1)$.
