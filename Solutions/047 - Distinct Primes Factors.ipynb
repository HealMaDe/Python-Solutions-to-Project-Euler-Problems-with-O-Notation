## Main Solution <a name="Main-Solution"></a>
First we try to define a function that generates prime factors of a given numbers. Note that I use a set to eliminate duplicate prime factors

```python
def prime_factors(n):
    factors_set = set()
    factor = 2

    while factor * factor <= n:
        if n % factor == 0:
            while n % factor == 0:
                n //= factor
            factors_set.add(factor)
        factor += 1

    if n > 1:
        factors_set.add(n)

    return factors_set
```

Then we tackle the main problem

```python
start_time = time.time()

n = 647
while True:
    if len(prime_factors(n)) == len(prime_factors(n + 1)) == 4:
      if len(prime_factors(n + 2)) == len(prime_factors(n + 3)) == 4:
        answer = n
        break
    n += 1

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Time Complexity:
   - Prime Factorization: The `prime_factors` function performs prime factorization on a given number `n`. The time complexity of this function is approximately $O(\sqrt{n})$, as it iterates up to the square root of `n` to find prime factors. This significantly reduces the number of iterations, resulting in a more efficient prime factorization process.
   - While Loop: The while loop in the main code iterates until it finds a number `n` that has four consecutive numbers (`n`, `n+1`, `n+2`, `n+3`) with exactly four prime factors. In the worst case, this loop iterates through a range of numbers until the condition is met. The exact time complexity for this loop depends on the input and cannot be determined precisely.

   Therefore, the overall time complexity of the code can be considered as $O(\sqrt{n})$ for prime factorization, with an additional time complexity for the while loop.

2. Space Complexity:
   - Factors Set: The `prime_factors` function creates a set to store the prime factors. The space complexity of this set depends on the number of distinct prime factors of `n`. In the worst case, if `n` has four distinct prime factors, the space complexity is $O(4)$ or $O(1)$.
   - Additional Variables: The code uses additional variables for iteration and temporary storage. These variables have a constant space complexity of $O(1)$.

   Therefore, the overall space complexity of the modified code is $O(1)$ or constant.
