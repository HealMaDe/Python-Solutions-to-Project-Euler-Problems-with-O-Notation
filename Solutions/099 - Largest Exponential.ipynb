## Main Solution <a name="Main-Solution"></a>

Suppose we want to comare these two:

$a^b , c^d$

Since this problem deals with large numbers, it is better to impose logarithm function to our numbers and compare these below numbers instead:

$log(a^b)$ and $log(c^d)$ ---> $b * log(a)$ and $d * log(c)$

```python
start_time = time.time()

with open('0099_base_exp.txt', 'r') as file:
    # Initialize variables for storing each column
    base = []
    exponent = []

    # Read each line in the file
    for line in file:
        # Split the line into columns using a delimiter
        columns = line.strip().split(',')  # Change the delimiter to ',' if it's comma-separated

        # Store the contents of each column in the respective variables
        base.append(columns[0])
        exponent.append(columns[1])

L = []
for i in range(0,len(base)):
  L.append(int(exponent[i])*math.log(int(base[i])))

answer = L.index(max(L)) +1

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


**Time Complexity**:

1. **Reading the File**: The time complexity of reading the file is dependent on the size of the file, which is represented as $O(n)$, where $n$ is the number of lines in the file.

2. **Storing the Columns**: During the file reading process, we iterate through each line and split it into columns. This operation has a time complexity of $O(1)$ since the number of columns is fixed (assuming a constant number of columns).

3. **Calculating the List $L$**: The loop that calculates the values in the list $L$ iterates over the length of the base list, which has $n$ elements (assuming the $base$ and $exponent$ lists have the same length). Inside the loop, the calculations involve simple mathematical operations. Therefore, the time complexity of this loop is $O(n)$.

4. **Finding the Maximum Value**: The `L.index(max(L))` operation has a time complexity of $O(n)$ since it iterates through the list $L$ to find the maximum value.

Overall, the **time complexity** of my code is $O(n)$, where $n$ is the number of lines in the file or the length of the base and exponent lists.

<br>

**Space Complexity**:

1. **Storing Columns**: The space complexity for storing the columns ($base$ and $exponent$ lists) is $O(n)$, where $n$ is the number of lines in the file. This is because you need to store the values of each column separately.

2. **List $L$**: The space complexity for the $L$ list is also $O(n)$ since it has the same number of elements as the $base$ and $exponent$ lists.

3. **Additional variables**: The other variables ($answer$, $line$, $columns$) used in my code have constant space complexity and do not depend on the input size.

Therefore, the overall **space complexity** of my code is $O(n)$, where $n$ is the number of lines in the file or the length of the base and exponent lists.
