## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from fractions import Fraction  #work with rational numbers
```

## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

n1 = n2 = n3 = n4 = 1

for i in range(10,50):
    for j in range(50,100):
        a = str(i)[0]
        b = str(i)[1]
        c = str(j)[0]
        d = str(j)[1]
        if int(b) == int(d) == 0:
            continue
        else:
            if a == c and int(d)!=0:
                if i/j == int(b)/int(d):
                    n1 *= i/j
            elif a == d and int(c)!=0:
                if i/j == int(b)/int(c):
                    n2 *= i/j
            elif b == c and int(d)!=0:
                if i/j == int(a)/int(d):
                    n3 *= i/j
            elif b == d and int(c)!=0:
                if i/j == int(a)/int(c):
                    n4 *= i/j


answer = Fraction.from_float(n1*n2*n3*n4).limit_denominator().denominator

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the nested loops ($10$ to $49$ and $50$ to $99$). This is because the code iterates through all possible combinations of $i$ and $j$ within these ranges.

The **space complexity** of my solution is **$O(1)$**, because it only uses a constant amount of additional space to store the variables $n_1$, $n_2$, $n_3$, $n_4$, $a$, $b$, $c$, and $d$. The space used does not depend on the input size.
