## Main Solution <a name="Main-Solution"></a>

python sets are more quicker than lists

```python
start_time = time.time()

s = set()
for d in range(2, 12000 + 1):
    for n in range(d//3 + 1, d//2 + 1):
        if n/d > 1/3:
            if n/d < 1/2:
                s.add(n/d)
answer = len(s)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
1. Nested loops: The code contains nested loops that iterate over the range of values for $d$ and $n$. The number of iterations for the outer loop is $(12000 - 2 + 1)$, which is $O(1)$. For the inner loop, the number of iterations depends on the value of $d$. It iterates from `d//3 + 1` to `d//2 + 1`, which is approximately `(1/6 * d)` iterations. Therefore, the time complexity of the inner loop is $O(d)$.

2. Set operations: The code performs set operations to add elements to the set $s$. The time complexity of adding an element to a set is $O(1)$.

Therefore, the overall **time complexity** of the code is dominated by the nested loops, which is $O(d)$, where $d$ is the range from $2$ to $12000$.

Space Complexity:
1. Set: The code uses a set called $s$ to store unique values. The size of this set depends on the number of unique elements added to it. In the worst case, the number of unique elements is approximately $(12000 - 2 + 1) * (1/6 * d)$, which is $O(d^2)$. Therefore, the space complexity of the set $s$ is $O(d^2)$.

2. Other variables: The code uses additional variables like $d$ and $n$, which store integer values. These variables require a constant amount of space, so the space complexity of these variables is $O(1)$.

Therefore, the overall **space complexity** of the code is dominated by the set $s$, which is $O(d^2)$ in terms of the range from $2$ to $12000$.
