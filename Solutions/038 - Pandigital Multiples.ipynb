## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

a = 0
# Iterate through different integers
for i in range(1, 10000):
    seq = ""

    # Generate the concatenated product
    for n in range(1, 10):
        seq += str(i * n)

        if len(seq) == 9:
            # Check if the concatenated product is pandigital
            if "".join(sorted(seq)) == "123456789":
                a = max(a, int(seq))
            break
        elif len(seq) > 9:
            break

answer = a

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the range of integers being iterated through (in this case, $1$ to $10000$). This is because there are two nested loops - the outer loop iterates through the range of integers, and the inner loop iterates through the range of numbers to generate the concatenated product.

The **space complexity** of my solution is **$O(1)$**, as there are no data structures being used that grow with the input size. The only variable being used is $a$, which stores the largest pandigital number found so far
