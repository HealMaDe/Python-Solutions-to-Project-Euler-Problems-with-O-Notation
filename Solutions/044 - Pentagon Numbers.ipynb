## Main Solution <a name="Main-Solution"></a>
First I try to disciver charectristics of any number that can be pentagon.
the equation is:
$p = n(3n-1)/2$

based on this equation, $n$ is equal to $(\sqrt{1 + 24*p} + 1)/6$.

Since the term $n$ has to be an integer, we can conclude 2 important things:


1.   the term $\sqrt{1 + 24*p}$ , has to be perfect square and,
2.   the final answer of $(\sqrt{1 + 24*p} + 1)$ has to be divisible to $6$.

So based on above descriptions, first we define a function that checks if any given number is pentagonal or not.

```python
def is_pentagonal(n):
    # Check if a number is pentagonal
    p = ((24 * n + 1) ** 0.5 + 1) / 6
    return p == int(p)
```

Now we can tackle our problem.

```python
start_time = time.time()

minimum = float('inf')
for j in range(1, 10000):
    pj = j*(3*j - 1) // 2

    for k in range(j - 1, 0, -1):  #avoid unnecessary iterations
        pk = k * (3 * k - 1) // 2

        D = abs(pj - pk)

        if D >= minimum:
            break  # No need to continue searching for smaller differences

        if is_pentagonal(D) and is_pentagonal(pj + pk):
            minimum = D

answer = minimum

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the input number. This is because there are two nested loops that iterate from $1$ to $n$, resulting in a total of $n^2$ iterations.

The **space complexity** of my solution is **$O(1)$**, because it only uses a constant amount of additional space to store variables and does not depend on the size of the input.
