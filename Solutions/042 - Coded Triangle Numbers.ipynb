## Main Solution <a name="Main-Solution"></a>
First I try to disciver charectristics of any number that can be triangle.
the equation is:
$t = n(n+1)/2$

based on this equation, $n$ is equal to $(\sqrt{1 + 8*t} - 1)/2$.

Since the term $n$ has to be an integer, we can conclude 2 important things:


1.   the term $\sqrt{1 + 8*t}$ , has to be perfect square and,
2.   the final answer of $(\sqrt{1 + 8*t} - 1)$ has to be even.

So first we define a function that checks if any given number is perfect square or not.

```python
def perfect_square(n):
    sqrt = int(n**0.5)
    if sqrt*sqrt == n:
        return True
    else:
        return False
```

Then based on description above, we define another function to check if any given number could be a triangle number or not.

```python
def triangle_number(n):
    if perfect_square(1 + 8*n):  #must be perfect square
        if ((1 + 8*n)-1)%2 == 0: #must be even
            return True
    return False
```

Now we can tackle our problem by opening the file and give each letter its proper values.

```python
start_time = time.time()

alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
d = {}
for i in range(len(alphabet)):
    d[alphabet[i]] = i + 1

with open("42_words.txt", 'r') as file:
    content = file.read()

# Split the words by comma delimiter and remove quotation marks
words = content.replace('"', '').split(',')

count = 0
for word in words:
    sum = 0
    for i in word:
        sum += d[i]

    if triangle_number(sum):
        count += 1

answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the number of words in the file. This is because for each word, we iterate through each character in the word to calculate the sum of their corresponding values. Additionally, we check if the sum is a triangle number, which involves calculating the square root of the sum. Both of these operations have a time complexity of $O(n)$, resulting in an overall time complexity of $O(n^2)$.

The **space complexity** of my solution is **$O(1)$**, because we are not using any additional data structures that grow with the input size. The only space used is for the dictionary, which has a fixed size of 26 elements.
