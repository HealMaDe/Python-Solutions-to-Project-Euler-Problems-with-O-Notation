## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from collections import defaultdict  #building a dictionary that its default values are list
```

## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

cubes = defaultdict(list)
n = 1   #starting number

while True:
  cube = n ** 3
  digits = "".join(sorted(str(cube)))

  cubes[digits].append(cube)
  if len(cubes[digits])==5:
    answer = min(cubes[digits])
    break
  n += 1

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code consists of a `while True` loop that continues until a certain condition is met. In each iteration of the loop, the code performs a fixed set of operations.
- The key operations within each iteration include computing the cube (`cube = n ** 3`), sorting the digits of the cube (`digits = "".join(sorted(str(cube)))`), and appending the cube to the `cubes` dictionary (`cubes[digits].append(cube)`).
- The time complexity of computing the cube (`n ** 3`) and sorting the digits (`"".join(sorted(str(cube)))`) is $O(log(cube)^2)$ since the number of digits in the cube is proportional to $log(cube)$, and sorting the digits takes an additional $O(log(cube))$ time.
- Appending the cube to the `cubes` dictionary (`cubes[digits].append(cube)`) has an average time complexity of $O(1)$ since it uses a defaultdict with lists as values.
- The loop continues until a condition is met, which is when `len(cubes[digits]) == 5` for a specific set of digits. The number of iterations required to reach this condition can vary depending on the input.

Overall, the **time complexity** of the code can be considered as $O(N*log(N)^2)$, where $N$ represents the number of iterations until the condition is met.

Space Complexity:
- The code uses a `defaultdict` named `cubes` with lists as values. The space required for this dictionary will depend on the number of unique sets of digits encountered during the execution of the loop.
- The space complexity of the defaultdict is proportional to the number of unique sets of digits encountered multiplied by the average number of cubes associated with each set of digits.
- In the worst case scenario, if the loop runs indefinitely without finding a set of digits with 5 cubes, the space complexity will continue to grow.
- Additionally, there is space required for the variables `n`, `cube`, and `digits`, which are of constant size.

In summary, the $space complexity$ of the code can be considered as $O(M * K)$, where $M$ represents the number of unique sets of digits encountered and $K$ represents the average number of cubes associated with each set of digits.
