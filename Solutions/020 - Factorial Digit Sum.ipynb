## Main Solution <a name="Main-Solution"></a>
First We define a function that efficiently calculate factorial of a number.
Please note that I did not use any recursive function, as It requires a massive memory to run. instead I use a list. If you want to use recursive function at any cost, I strongly recommend to consider [memoization technique](https://en.wikipedia.org/wiki/Memoization#:~:text=In%20computing%2C%20memoization%20or%20memoisation,the%20same%20inputs%20occur%20again.) as I used it in Problem 14.

```python
def fact(n):
    L = [1]
    for i in range(1,n+1):
        L.append(L[-1]*i)
        del L[0]
    return L[-1]
```

```python
start_time = time.time()

L1 = []
s = str(fact(100))
for i in range(0,len(s)):
    L1.append(int(s[i]))

answer = sum(L1)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the input number. This is because the code iterates $n$ times in the for loop to calculate the factorial.

The **space complexity** of my solution is **$O(1)$**, because the space used by the code does not depend on the input size. The code only uses a constant amount of space to store the factorial and the sum of the digits.
