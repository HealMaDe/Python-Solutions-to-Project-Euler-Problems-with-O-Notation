## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

def answer():
  p = [1]  # List to store partition numbers
  n = 1    # Starting value of n

  while True:
      next_p = 0  # Variable to store the next partition number
      k = 1       # Counter for pentagonal numbers
      sign = 1    # Sign to alternate between addition and subtraction

      while True:
          p1 = k * (3 * k - 1) // 2  # k-th pentagonal number (1)
          p2 = k * (3 * k + 1) // 2  # k-th pentagonal number (2)

          if p1 > n:
              break  # Break if pentagonal number (1) exceeds n

          next_p += sign * p[n - p1]  # Add or subtract the corresponding partition number
          next_p %= 10**6             # Take modulo 1 million

          if p2 > n:
              break  # Break if pentagonal number (2) exceeds n

          next_p += sign * p[n - p2]  # Add or subtract the corresponding partition number
          next_p %= 10**6             # Take modulo 1 million

          k += 1      # Increment k for the next pentagonal number
          sign *= -1  # Alternate the sign between addition and subtraction

      if next_p == 0:
          return n    # If the next partition number is divisible by one million we got the answer

      p.append(next_p)  # Add the next partition number to the list
      n += 1            # Increment n for the next iteration

answer = answer()
end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The outer loop iterates indefinitely until the condition is explicitly broken.
- The inner loop iterates until the condition is explicitly broken.
- The inner loop's break conditions are based on the comparison of pentagonal numbers with `n`.
- The calculation of pentagonal numbers `p1` and `p2` is constant time.
- The addition and subtraction of partition numbers is also constant time.
- The modulo operation `next_p %= 10**6` is constant time.
- The increment of `k` and the alternation of `sign` are constant time.

Considering the worst case scenario where the inner loop iterates until both `p1` and `p2` exceed `n`, the number of iterations can be approximated by the square root of `n`. The inner loop can be considered to have a time complexity of $O(\sqrt{n})$.

As a result, the overall time complexity of the code is approximately $O(\sqrt{n})$ in the worst case scenario.

Space Complexity:
- The list `p` stores the partition numbers, which grows with each iteration.
- The size of the list `p` is directly proportional to the value of `n`.
- The space complexity is determined by the maximum size of the list `p`, which is reached when the condition for printing `n` is met.
- Therefore, the space complexity is $O(n)$.

In summary, the code has a **time complexity** of approximately $O(\sqrt{n})$ and a **space complexity** of $O(n)$.
