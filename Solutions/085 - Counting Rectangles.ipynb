## Main Solution <a name="Main-Solution"></a>

Let's analyse the problem a little bit. If you think about rectangle in a grid, It seems that we need 2 lines in a vertical axis and 2 lines in a horizontal axis. Now Suppose that we have a grid with length of $x$ and width of $y$. This means we have $(x+1)$ horizontal lines and $(y+1)$ vertical lines. So there are $C(x,2)$ ways to pick horizontal lines and $C(y,2)$ ways to pick vertical lines.

```python
start_time = time.time()

minimum = float("inf")
L = [0]
for x in range(1,500):
    for y in range(x,500):
        if abs((x*(x+1)*y*(y+1)//4 - 2*10**6)) < minimum:
            minimum = abs((x*(x+1)*y*(y+1)//4 - 2*10**6))
            L[0]=(x,y)

answer = L[0][0]*L[0][1]

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
1. Outer loop: The outer loop iterates from $x = 1$ to $499$, which takes $O(1)$ time.

2. Inner loop: The inner loop iterates from $y = x$ to $499$, which takes $O(n)$ time, where $n$ is the difference between the starting and ending values of $y$, which is approximately $499 - x$.

3. Calculation and Comparison: Inside the inner loop, there are constant time operations to calculate the expression `x*(x+1)*y*(y+1)//4 - 2*10**6` and compare it with the current minimum value. These operations take $O(1)$ time.

Therefore, the overall **time complexity** of the code is roughly $O(1 + (499-1) + (498-2) + (497-3) + ... + (2-498) + (1-499))$, which simplifies to $O(249,001)$. This can also be approximated as $O(1)$ since the loop iterations are fixed and do not depend on the input size.

Space Complexity:
1. Variables: The code uses a constant number of variables (`minimum`, `L`, `x`, `y`) that require $O(1)$ space.

Therefore, the overall **space complexity** of the code is $O(1)$.
