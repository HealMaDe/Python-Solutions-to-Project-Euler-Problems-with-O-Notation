## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

a,b = 1,3
total = 0
limit = 10**9
while True:
    perimeter = 12*a*a+4
    if perimeter > limit:
      break
    total += perimeter
    perimeter = 6*(b*b) - 4
    if perimeter > limit:
      break
    total += perimeter
    a , b = a + b , 2*a + 3*b

answer = total

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
The code consists of a while loop that continues until a condition is met. Within the while loop, there are a series of constant-time operations. Let's analyze the time complexity of each step:

1. Assignment Operations: The assignments `a, b = a + b, 2 * a + 3 * b` involve basic arithmetic operations and take constant time.

2. Perimeter Calculation: The calculations `perimeter = 12 * a * a + 4` and `perimeter = 6 * (b * b) - 4` involve multiplication and addition operations, which take constant time.

3. Comparison and Break Conditions: The comparisons `perimeter > limit` and the associated break statements depend on the value of `perimeter` and `limit` and take constant time.

The while loop continues until the break conditions are met. The number of iterations depends on the values of `a` and `b` and how quickly the perimeter exceeds the limit. The exact number of iterations is not known without further analysis.

Overall, the **time complexity** of the code can be approximated as $O(N)$, where $N$ is the number of iterations until the break conditions are met. The exact value of N depends on the specific values of `a`, `b`, and the limit.

Space Complexity:
The space complexity of the code is determined by the variables used. In this case, there are four variables: `a`, `b`, `total`, and `limit`. Each of these variables requires a constant amount of space.

Therefore, the **space complexity** of the code is $O(1)$, as it uses a fixed amount of memory regardless of the input size.
