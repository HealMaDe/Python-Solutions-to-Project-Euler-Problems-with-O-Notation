## Main Solution <a name="Main-Solution"></a>

1.   $a$ & $b$ even:

  *     $a^2 + b^2 = c^2$  ---> $c^2$ is even ---> $c$ is even   
  *     Since $p = a + b + c$ ---> $p$ is even too


2.   $a$ & $b$ odd:

  *     $a^2 + b^2 = c^2$  ---> $c^2$ is even ---> $c$ is even   
  *     Since $p = a + b + c$ ---> $p$ is even too

3.   $a$ even & $b$ odd or vice versa:

  *     $a^2 + b^2 = c^2$  ---> $c^2$ is odd ---> $c$ is odd   
  *     Since $p = a + b + c$ ---> $p$ is even

So the parameter $p$ is always even. That's a good thing to know.
Now let's proceed and try to write parameters only based on $p$ and $a$:

$c = p - a - b $

$ a^2 + b^2 = (p - a - b)^2$

if we calculate $b$, we get:

$b = p(2a-p)/2(a-p)$

Note that $b$ has to be an integer. That's it. Let's code based on what we've learned about our parameters.




```python
start_time = time.time()

d = {}
for p in range(2, 1000, 2):
    count = 0
    for a in range(1, p // 3):  # Limiting the range of 'a' based on the conditions
        if p * (2*a - p) % (2*(a - p)) == 0:
            count += 1
    d[p] = count


answer = max(d, key=d.get)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the number of even numbers from $2$ to $1000$. This is because there is a nested loop where the outer loop iterates over the even numbers from $2$ to $1000$, and the inner loop iterates from $1$ to $p // 3$, which is a fraction of the outer loop variable $p$. Therefore, the number of iterations in the inner loop is dependent on the value of $p$, resulting in a quadratic time complexity.

The **space complexity** of my solution is **$O(n)$**, where $n$ is the number of even numbers from $2$ to $1000$. This is because the dictionary $d$ stores the count values for each even number, resulting in a space complexity proportional to the number of even numbers.
