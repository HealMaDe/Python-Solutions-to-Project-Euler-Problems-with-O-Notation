## Main Solution <a name="Main-Solution"></a>
First we generate all primes between two given numbers based on [Sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes) Algorithm.

```python
def generate_primes_between(m,n): # m < n
    sieve = [True]*(n + 1)
    sieve[0] = sieve[1] = False  #0 and 1 are not primes
    p = 2
    while p**2<n:
        for i in range(p**2,n+1,p):
            sieve[i] = False

        p += 1
    primes = [p for p in range(len(sieve)) if sieve[p] and p>m]

    return primes
```

Then we tackle the main problem

```python
start_time = time.time()

primes = generate_primes_between(1488,9999)

for i in primes:
    if i + 3330 in primes and i + 2*3330 in primes:
        if sorted(str(i)) == sorted(str(i + 3330)) == sorted(str(i + 2*3330)):
            answer = str(i)+ str(i + 3330) + str(i + 2*3330)


end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


**Time Complexity**:

1. **Generating Primes**: The function uses the Sieve of Eratosthenes algorithm to generate prime numbers up to $n$. The time complexity of this function is approximately $O(n*log(log(n)))$ as it involves marking composite numbers in the sieve. This step has a significant impact on the overall time complexity.

2. **Loop Iterations**: The loop that iterates over the generated prime numbers has a linear time complexity. However, it is important to note that the number of prime numbers between $m$ and $n$ is significantly smaller than $n$, so the impact on the overall time complexity is relatively small.

3. **Sorting and Comparison**: Sorting the digits of the numbers and the subsequent comparisons have a time complexity of $O(k*log(k))$, where $k$ represents the number of digits in the numbers. Since the numbers in my code have a fixed number of digits (4 digits), the time complexity can be considered constant.

Therefore, the overall **time complexity** of my code is dominated by the Sieve of Eratosthenes algorithm, resulting in approximately $O(n*log(log(n)))$.


<br>

**Space Complexity**:

1. **Sieve**: The function creates a sieve list of size $n+1$. Hence, the space complexity of the sieve is $O(n)$.

2. **primes List**: The function also creates a primes list to store the generated prime numbers between $m$ and $n$. The size of this list depends on the number of prime numbers between $m$ and $n$, which is significantly smaller than $n$. Therefore, the space complexity of the primes list can be considered relatively small.

Overall, the **space complexity** of my code is $O(n)$ due to the sieve list. The space complexity of the primes list is relatively small compared to $n$ in this case.
