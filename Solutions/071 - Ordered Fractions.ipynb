## Main Solution <a name="Main-Solution"></a>

We want the fraction right before $3/7$ , right?

Let me play with the fraction $3/7$ a bit.

$3/7 = 3*P / 7*P$

when $P$ is large, the fraction right before $3/7$ is equal to $(3*P-1)/7*P$

So All we have to do is find $P$ based on our limitation on $d$ in problem. Then the answer of the problem is equal to $3P - 1$ :)

```python
start_time = time.time()

limit = 10**6
d = 1
while 7*d<10**6:
  d += 1

P = d - 1
answer = 3*P - 1

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code consists of a single while loop that increments the value of `d` until the condition `7 * d < 10 ** 6` is no longer satisfied.
- The while loop runs iterations until the condition is met, which occurs when `d` reaches a value such that `7 * d >= 10 ** 6`.
- Therefore, the number of iterations in the while loop can be approximated as `10 ** 6 / 7`, resulting in a time complexity of $O(1)$ since the number of iterations is fixed.

- The subsequent operations `P = d - 1` and `answer = 3 * P - 1` are constant time operations, resulting in $O(1)$ time complexity.

- Overall, the time complexity of the code is $O(1)$ since the number of iterations in the while loop is fixed and the subsequent operations are constant time.

Space Complexity:
- The code initializes a single variable `d`, which requires constant space.

- The variables `P` and `answer` also require constant space.

- Therefore, the space complexity of the code is $O(1)$ since it uses a fixed amount of space regardless of the input size.
