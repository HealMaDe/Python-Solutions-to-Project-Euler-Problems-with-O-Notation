## Problem 88 <a name="Problem-88"></a>

<br>
$**Hint**$

Find the sum of all minimal numbers of the form $N = a_1 + ... + a_k$
for which $N = a_1 * ... * a_k$ as well, where $k$ varies from $2$ to $12000$.

## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
```

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

numbers = {k: -1 for k in range(2, 12001)}

def Factorizations(n, k=2):
    result = []
    while k * k <= n:
        if n % k == 0:
            # Recursively find factorizations of n/k starting from k
            result.extend([[k] + f for f in Factorizations(n // k, k)])
        k += 1
    return result + [[n]]

# Iterate over numbers from 2 to 19999
for n in range(2, 20000):
    for f in Factorizations(n):
        i, j = sum(f), len(f)
        if i <= n:
            k = (n - i) + j
            # Check conditions for k to be within range and not already updated
            if 2 <= k <= 12000 and numbers[k] == -1:
                numbers[k] = n

# Sum all unique values in minnumbers except for -1
answer = sum(set(n for k, n in numbers.items() if n != -1))


end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
1. Initializing `numbers` dictionary: The initialization of `numbers` has a time complexity of $O(1)$ since it involves creating a dictionary with a constant number of elements.

2. `Factorizations` function: The time complexity of this function depends on the input `n`. In the worst case, the function performs factorization by trial division, which has a time complexity of approximately $O(\sqrt{n})$. However, since the function calls itself recursively for each factor found, the overall time complexity can be considered as $O(\sqrt{n} * m)$, where `m` is the number of distinct prime factors of `n`.

3. Outer loop (range 2 to 19999): This loop iterates over numbers from 2 to 19999. Therefore, the time complexity of this loop is $O(n)$, where `n` is the range size.

4. Inner loop (Factorizations): The inner loop iterates over the factorizations obtained from the `Factorizations` function. The number of factorizations for a given `n` can vary, but in the worst case, it can be considered as $O(\sqrt{n} * m)$, as explained earlier.

Overall, the **time complexity** of the code can be approximated as $O(n * \sqrt{n} * m)$, where `n` is the range size and `m` represents the number of distinct prime factors within the range.

Space Complexity:
1. `numbers` dictionary: The space complexity of the `numbers` dictionary is $O(1)$ since it contains a constant number of elements.

2. `Factorizations` function: The space complexity of the `Factorizations` function is determined by the recursive calls and the `result` list. In the worst case, the function can have a space complexity of $O(\sqrt{n} * m)$, where `n` is the input number and `m` is the number of distinct prime factors.

Overall, the **space complexity** of the code can be approximated as $O(\sqrt{n} * m)$, where `n` is the input number in the `Factorizations` function and `m` represents the number of distinct prime factors within the range.
