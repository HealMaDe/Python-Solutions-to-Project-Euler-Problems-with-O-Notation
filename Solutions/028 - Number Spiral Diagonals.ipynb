## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

period = 4  #numbers at the edge have the same step for 4 times. except starting number
answer = 1  #sum of diagonal numbers
diff = 2    #diffference between 2 desired number at start
n = 1  #starting number

while n != 1001*1001:
    for i in range(period):
        n += diff
        answer += n
    diff += 2

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the value of the variable $n$ when the while loop terminates. This is because the while loop will iterate until $n$ reaches $1001*1001$, and the number of iterations is directly proportional to the value of $n$.

The **space complexity** of my solution is **$O(1)$**, as it only uses a
fixed amount of memory to store the variables $period$, $answer$, $diff$, and $n$. The amount of memory used does not depend on the input size.
