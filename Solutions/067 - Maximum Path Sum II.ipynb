## Main Solution <a name="Main-Solution"></a>

Insted of starting at the top of the pyramid and moving to adjacent numbers on the row below, Do the exact opposite (bottom to up)

```python
start_time = time.time()

#opening the file and read it
file = open("067_triangle.txt","r")

#building our pyramid list to work with
pyramid = []
for line in file:
  line = line.strip()
  row = [int(i) for i in line.split() ]
  pyramid.append(row)

#bottom up approach
while len(pyramid)!= 1:
    for i in range(len(pyramid[-2])):
        pyramid[-2][i] += max(pyramid[-1][i],pyramid[-1][i+1])
    del pyramid[-1]

answer = pyramid[0][0]

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the number of rows in the pyramid. This is because we iterate through each element in the pyramid, and for each element, we perform a constant number of operations (comparing and adding values). Since the pyramid has $n(n+1)/2$ elements, the overall time complexity is $O(n^2)$.

The **space complexity** of my solution is **$O(1)$**, as we are modifying the original pyramid list in-place and not using any additional data structures that scale with the input size.
