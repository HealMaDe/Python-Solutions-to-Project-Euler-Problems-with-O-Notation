## Main Solution <a name="Main-Solution"></a>

If wa want to have $n$-digit for number $x^n$, then the following relation must be valid:

$10^{n-1} <= x^n < 10^n$

if we simplify above inequalities we get:

$10^{(n-1)/n} <= x < 10$

so now we have an upper and lower limit for $x$. Let's proceed and see what sort of thing we can have for the values of $n$.

Since $10^{(n-1)/n}$ has to be less than $10$ we can conclude that n must be integers that makes the term $10^{(n-1)/n}$ less than $10$.
if we calulat the limit of $10^{(n-1)/n}$ when n is equal to infinity, we get the number 10 that could not be valid. so we have to find this infinity n integer which for that our inequality would have been violated. It turns out that n would be 22. this means after n = 22, the term $10^{(n-1)/n}$ stucks at the decimal number $9.n_1n_2n_3n_4....$ till infinity. So we can conclude that if the term $10^{(n-1)/n}$ equals to 9, our job is done.


```python
start_time = time.time()

count = 0
n = 1
found = True
while found:
    lower_limit = int(10**((n - 1) / n))
    upper_limit = 10

    for x in range(lower_limit, upper_limit):
        power = x ** n
        if len(str(power)) == n:
            count += 1
        if lower_limit == 9:
            found = False
            break
    n += 1

answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Time Complexity:
   - Iterative Loop: The code has an iterative loop that starts from `n = 1` and continues until the condition `found` is set to `False`. The number of iterations depends on the value of `n` and the condition within the loop. Let's denote the final value of `n` as `N`. The time complexity of this loop can be approximated as $O(N)$.
   - Nested Loop: Within the iterative loop, there is a nested loop that iterates over the values of `x` within the range of the lower limit and upper limit. The number of iterations depends on the difference between the upper and lower limits for each `n`. The time complexity of this nested loop can be approximated as $O(upper limit - lower limit)$.

   Therefore, the overall **time complexity** can be estimated as the sum of the time complexities of the iterative loop and the nested loop: $O(N * (upper limit - lower limit))$.

2. Space Complexity:
   - Variables: The code uses a few variables to keep track of the count, the current value of `n`, the lower limit, the upper limit, and the boolean `found`. The space complexity of these variables is constant and does not depend on the input size.

   Therefore, the overall **space complexity** is constant: $O(1)$.
