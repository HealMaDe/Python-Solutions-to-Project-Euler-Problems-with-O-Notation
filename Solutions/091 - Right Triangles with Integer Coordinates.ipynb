## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

answer = 0
for x1 in range(0,51):
    for x2 in range(x1,51):
        for y1 in range(0,51):
            for y2 in range(0,y1+1):

                a = (x1**2 + y1**2)**0.5
                b = (x2**2 + y2**2)**0.5
                c = ((x2-x1)**2 + (y2-y1)**2)**0.5

                if min(a,b,c) == 0:
                    continue
                cache = max(a,b,c)

                if cache == a:
                    if round(b**2 + c**2,10) == round(a**2,10):
                        answer += 1

                elif cache == b:
                    if round(a**2 + c**2,10) == round(b**2,10):
                        answer += 1

                else:
                    if round(a**2 + b**2,10) == round(c**2,10):
                        answer += 1

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

The time and space complexity of my code can be analyzed as follows:

Time Complexity:

The code includes four nested loops that iterate over the ranges 0 to 50. Therefore, the time complexity can be approximated as $O(n^4)$, where $n$ is the range limit (51 in this case). This means that the execution time of the code will increase rapidly as the range limit increases.

Within the loops, the code performs some mathematical computations, such as square roots and rounding. These mathematical operations typically have a time complexity of $O(1)$ because they operate on individual numbers rather than on a collection of data.

Overall, the dominant factor in the time complexity of the code is the four nested loops.

Space Complexity:

The space complexity of the code is relatively low because it does not use any significant additional data structures that grow with the input size. The code only uses a single variable `answer` to store the result.

The mathematical computations performed in the code do not require additional space proportional to the input size. They only involve simple arithmetic operations on individual numbers.

Therefore, the space complexity of the code can be considered as $O(1)$ or constant space.
