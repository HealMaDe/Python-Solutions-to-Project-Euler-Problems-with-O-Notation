## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()


# a = multiplicand , b = multiplier

#case 1
# multiplicand: 1 digit, multiplier: 4 digit, product: 4 digit
c = []
for a in range(2,9):
    for b in range(1234,9876):
        if a*b <= 9876 and sorted(str(a) + str(b) + str(a*b)) == ["1","2","3","4","5","6","7","8","9"]:
            c.append(a*b)


#case 2
# multiplicand: 2 digit, multiplier: 3 digit, product: 4 digit

for a in range(12,98):
    for b in range(123,987):
        if a*b <= 9876 and sorted(str(a) + str(b) + str(a*b)) == ["1","2","3","4","5","6","7","8","9"]:
            c.append(a*b)


answer = sum(set(c))

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the range of values for $a$ and $b$. This is because there are two nested loops, one for iterating through the range of values for $a$ and the other for iterating through the range of values for $b$. Since both loops have a range of $n$, the total number of iterations is $n * n = n^2$.

The **space complexity** of my solution is **$O(1)$**, which means it requires constant space. This is because the variable $c$ is the only additional space used, and its size does not depend on the input size. It only stores the valid products found, and the size of the set $c$ is limited to the number of valid products, which is a constant value.
