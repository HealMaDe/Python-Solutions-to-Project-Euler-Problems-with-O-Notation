## Main Solution <a name="Main-Solution"></a>

$B$: Number of blue discs

$R$: Number of red discs

$T = B + R$ , Total discs

we want: $B(B-1)/T(T-1) = 1/2$

if we play with this equation and replace $B$ and $R$, we get the following equation for $B$:

$B = [1+2R+\sqrt{1 + 8R^2}]/2$

since we deal with integer numbers, the term $\sqrt{1 + 8R^2}$ has to be perfect square. that gives us something to start.

if we caculate the R numbers so that the above term is perfect square we get this series:
$1, 6, 35, 204, 1189, 6930, 40391, 235416, 1372105, 7997214 , ....$

If we look closely we can identify if we multiply each term of the sequence to 5.8284, we get the next term.

Now all we have to do is that to find $R$ such that $B+R$ exeeds 10**12


```python
start_time = time.time()

def is_perfect_square(n):
    sqrt = int(n**0.5)
    return sqrt*sqrt == n

r = 1
count = 0
L = []
while len(str(r))!= 13:
    if is_perfect_square(1 + 8*r**2):
        L.append(r)
        r  = int(r * 5.8284)

    r += 1

R = L[-1]
B = int((1 + 2*R + (1 + 8*R**2)**0.5)/2)
answer = B

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The `is_perfect_square` function has a time complexity of $O(1)$ because it performs a constant number of operations to calculate the square root and compare the result.
- The `while` loop continues until the length of the string representation of `r` is equal to 13. The number of iterations required to reach this condition depends on the value of `r`.
- Inside the loop, there are several operations that have a constant time complexity, such as checking if a number is a perfect square, appending `r` to the list `L`, and updating the value of `r`.
- The loop also includes the operation `r = int(r * 5.8284)`, which scales `r` by a constant factor.
- Therefore, the overall time complexity of the code can be considered as $O(N)$, where $N$ is the number of iterations of the `while` loop until the condition is met.

Space Complexity:
- The space complexity mainly depends on the list `L`, which stores the values of `r`.
- The size of the list `L` will vary depending on the number of iterations of the `while` loop until the condition is met.
- In the worst case, if the loop runs indefinitely without meeting the condition, the size of `L` will continue to grow.
- The space complexity for storing the list `L` can be considered as $O(N)$, where $N$ is the number of iterations of the `while` loop until the condition is met.
- Other variables like `sqrt`, `count`, `r`, `R`, `B`, and `answer` have constant space requirements.

In summary, the **time complexity** of the code is $O(N)$, and the **space complexity** is $O(N)$, where $N$ is the number of iterations of the `while` loop until the condition is met.
