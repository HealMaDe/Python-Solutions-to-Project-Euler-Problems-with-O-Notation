## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

coins = [1, 2, 5, 10, 20, 50, 100, 200]

target = 200
ways = [0] * (target + 1)
ways[0] = 1

for coin in coins:
    for i in range(coin, target + 1):
        ways[i] += ways[i - coin] #recursion line

answer = ways[target]

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time complexity:
- The code uses a nested loop structure to iterate over the coins and the target value.
- The outer loop iterates over the coins, which has a fixed length of 8 (the number of elements in the `coins` list).
- The inner loop iterates from the value of the current coin (`coin`) to the target value (`target + 1`).
- The number of iterations in the inner loop depends on the value of `coin` and the target value. In the worst case, the inner loop iterates `target + 1` times.
- Therefore, the overall time complexity of the nested loops is $O(8 * (target + 1))$ or simply $O(target)$.

- The line `ways[i] += ways[i - coin]` inside the inner loop has constant time complexity as it performs a simple addition operation.

- Overall, the **time complexity** of the code is approximately $O(target)$.

Space complexity:
- The code uses a list `ways` of size `target + 1` to store the number of ways to make change for each value from 0 to the target.
- Therefore, the space complexity of the `ways` list is O(target).

- The code also uses a few integer variables (`target`, `coin`, and `answer`), which require constant space.

- The **space complexity** of the code is dominated by the `ways` list, resulting in $O(target)$ space complexity.
