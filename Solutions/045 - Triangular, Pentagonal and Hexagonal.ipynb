## Main Solution <a name="Main-Solution"></a>
First I try to disciver charectristics of any number that can be triangular.
the equation is:

$T = n(n+1)/2$

based on this equation, $n$ is equal to $(\sqrt{1 + 8*T} - 1)/6$.

Then I try to disciver charectristics of any number that can be pentagonal.
the equation is:

$p = n(3n-1)/2$

based on this equation, $n$ is equal to $(\sqrt{1 + 24*p} + 1)/6$.

All hexagonal numbers, are also triangular.

Now Let me define our functions.

```python
def T(n):   #Triangular
    return 0.5*n*(n+1)

def P(n):   #Pentagonal
    return 0.5*n*(3*n-1)

def H(n):   #Hexagonal
    return n*(2*n-1)
```

Now we can tackle our problem. we start our loop over the n3 which we assume belong to H(n). Then when we come to n2 which is belong to P(n), we replace the P with the Hexagonal formula. we do this to n1 too.

```python
start_time = time.time()

for n3 in range(1,100000):
    n2 = int((1+ (1 + 24*n3*(2*n3-1))**0.5)/6)
    n1 = int((-1 + (1 + 8*n2*(3*n2-1)/2)**0.5)/2)
    if P(n2) == H(n3) and n1 > 285:
        answer = int(T(n1))

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^3)$**, because there are three nested loops: one for $n3$, one for $n2$, and one for $n1$. Each loop iterates up to $n$, so the overall time complexity is $O(n^3)$.

The **space complexity** of my solution is **$O(1)$**, because it only uses a constant amount of space to store the variables $n1$, $n2$, and $n3$.
