## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
day = ["Mondy","Tuesday","Wednesday","Thirsday"
       ,"Friday","Saturday","Sunday"]

count = 1
c = 0
for year in range(1901,2001):
    if year % 400==0:
        del month[1]
        month.insert(1,29)
    elif year % 4 == 0 and year % 100 != 0:
        del month[1]
        month.insert(1,29)

    for i in range(0,12):
        for j in range(0,month[i]):
            count += 1
            if j == 0 and count % 7 == 6:
                c += 1
answer = c

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the number of years from 1901 to 2000. This is because the code iterates over each year in the range and performs a constant number of operations for each year.

The **space complexity** of my solution is **$O(1)$**, as it only uses a fixed amount of memory to store the month and day arrays, as well as the $count$ and $c$ variables. The size of the input (number of years) does not affect the amount of memory used by the code.
