## Main Solution <a name="Main-Solution"></a>

First I define a function to distinguish perfect squares. Then we address the problem.

```python
start_time = time.time()

def is_square(n):
    sqrt = int(n**0.5)
    return sqrt*sqrt == n

count = 0
for N in range(2,10001):
    if is_square(N):
        continue


    a0 = int(N**0.5)
    L = [(1,a0)]
    while True:

        b = (N - L[-1][1]**2)/L[-1][0]
        mult = int((N**0.5 + L[-1][1])/b)
        a = b*mult - L[-1][1]
        t = (b,a)
        if t in L: #period
            break
        else:
            L.append(t)
    if len(L)%2 == 1:  #odd periods
        count += 1

answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code consists of a loop that iterates over the range from $2$ to $10,000$ (`for N in range(2, 10001)`).
- Inside this loop, there is a while loop that continues until a repeating pattern is found (or a break condition is met).
- The while loop iteratively calculates and appends values to the list `L`.
- The time complexity of the while loop depends on the number of iterations it takes to find the repeating pattern.
- In the worst case, the while loop may iterate up to a maximum of `N` times.
- Therefore, the overall **time complexity** of the code is $O(N^2)$ as the outer loop iterates $N$ times and the inner while loop has a maximum of $N$ iterations.

Space Complexity:
- The space complexity is determined by the space used to store the list `L` and a few additional variables.
- The size of the list `L` depends on the number of elements it contains, which is the period of the repeating pattern.
- In the worst case, the size of `L` can be as large as N (when there is no repeating pattern).
- Therefore, the **space complexity** of the code is $O(N)$ as the list `L` can grow up to N elements.

Overall, the **time complexity** of the code is $O(N^2)$ and the **space complexity** is $O(N)$.
