**$Hint$**

We can use the Sieve of Eratosthenes algorithm. This algorithm allows us to generate a list of prime numbers up to a given limit. Here are the steps to implement the algorithm:

1. Create a boolean array of size $N+1$, where $N$ is the given limit, initialized with True values. This array will be used to mark numbers as prime or non-prime.

2. Mark the first two elements ($0$ and $1$) as False, as they are not prime.

3. Start with the first prime number, $2$.

4. Iterate through the array from $2$ to the square root of $N$. For each prime number found at index $p$, check if it is marked as True. If so, it means it is prime.

5. For each prime number $p$, iterate through the array starting from $p^2$ up to $N$, incrementing by $p$ at each step. Mark each multiple of $p$ as False, as they are not prime. Note that multiples of $p$ less than $p^2$ would have already been marked as False by previous prime numbers.

6. Repeat steps 4 and 5 until reaching the square root of $N$.

For further reading about this algorithm, check [here](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes).

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

def sum_primes(n):
    primes = [True] * (n + 1) #create a list of True values in length on n+1
    primes[0] = primes[1] = False  # 0 and 1 are not prime

    p = 2
    while p**2 <= n:
        if primes[p]:
            for i in range(p**2, n + 1, p):
                primes[i] = False
        p += 1

    summation = 0
    for i in range(2, n + 1):
        if primes[i]:
            summation += i

    return summation

n = 2000000
answer = sum_primes(n)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n *log(log(n))$**. This is because The algorithm iterates through all numbers from $2$ to $\sqrt{n}$  , marking all multiples of each number as non-prime. This process takes $O(n *log(log(n))$ time.

The **space complexity** of my solution is **$O(n)$**. This is because the function creates a list of size $(n+1)$ to store the boolean values indicating whether each number is prime or not.
