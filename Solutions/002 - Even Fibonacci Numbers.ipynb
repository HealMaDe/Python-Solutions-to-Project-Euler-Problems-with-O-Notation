## Main Solution <a name="Main-Solution"></a>

We can define the fibonacci function in 2 ways.


1.   Using Recursive function

```
def Recursive_fib(n):      # Fibonacci function
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return Recursive_fib(n-2)+Recursive_fib(n-1)
```


2.   More Efficient Way


```
def fib(n):
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        L = [1,2]
        for i in range(2,n):
            L.append(L[-1]+L[-2])

    return L[-1]
```





### Recursive Implementation <a name="Recursive-Implementation"></a>

````python
def Recursive_fib(n):      # Fibonacci function
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return Recursive_fib(n-2)+Recursive_fib(n-1)
````

````python
start_time = time.time()
n = 1
while Recursive_fib(n)<4000000:   # finding the last valid term
    n += 1

L = []
for i in range(2,n,3): #even valued terms
    L.append(Recursive_fib(i))

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:", sum(L),"\nExecution time:", "{:.5f}".format(execution_time), "seconds")
````

### More Efficient Implementation <a name="More-Efficient-Implementation"></a>

````python
def fib(n):
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        L = [1,2]
        for i in range(2,n):
            L.append(L[-1]+L[-2])

    return L[-1]
````

````python
start_time = time.time()
n = 1
while fib(n)<4000000:   # finding the last valid term
    n += 1

L = []
for i in range(2,n,3): #even valued terms
    L.append(fib(i))

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:", sum(L),"\nExecution time:", "{:.5f}".format(execution_time), "seconds")
````

As you can see the second implementation was faster. But the both method works quite good for this specific problem. but on a larger scale, the recursive method probably encounter some memory capacity issues.

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution for this problem in both implementation is **$O(n)$**, where n is the last valid term of the Fibonacci sequence thet is less then $4,000,000$. This is because the code uses a while loop to find the last valid term, which takes $O(n)$ iterations.

The **space complexity** of my solution for this problem in both implementation is also **$O(n)$**, where n is the last valid term of the Fibonacci sequence that is less than $4,000,000$. This is because the code uses a list to store the even-valued terms, which can have a maximum length of $n/3$.
