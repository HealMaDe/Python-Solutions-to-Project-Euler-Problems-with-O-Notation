## Main Solution <a name="Main-Solution"></a>

if $a <= b <=c$ so $c$ should always equal to $M$

```python
start_time = time.time()

count = 0
M = 2
while count <= 10**6:
    M += 1
    c = M  # a<=b<=c so c should always equal to M
    for ab in range(2, 2* M + 1):  # ab represents (a + b)
        if ((ab**2 + c**2) ** 0.5) % 1 == 0:
            if ab >= c:
                count += c - int((ab + 1) / 2) + 1
            else:
                count += int(ab / 2)

answer = M

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of my code can be analyzed as follows:

The outer `while` loop continues until the `count` variable exceeds $10^6$. The number of iterations in this loop depends on the value of `M` and how quickly the `count` variable increments.

In each iteration of the `while` loop, there is an inner `for` loop that iterates from $2$ to $2 * M + 1$. This loop has a constant number of iterations, specifically $(2 * M + 1) - 2$, which simplifies to $2 * M - 1$.

Inside the inner `for` loop, there are some constant time operations such as arithmetic calculations and condition checks.

Therefore, the overall **time complexity** of the code is approximately $O(M)$, where M is the value at which the `while` loop terminates.

As for the **space complexity**, it is constant throughout the code. There are no data structures or variables that depend on the input size. Thus, the space complexity is $O(1)$.
