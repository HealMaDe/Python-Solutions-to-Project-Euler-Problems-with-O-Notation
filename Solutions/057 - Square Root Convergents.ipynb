## Main Solution <a name="Main-Solution"></a>

the sequence of numbers are $3/2$, $7/5$, $17/12$, $41/29$, and so on. If we look closely there is a tricky pattern.

1. If we add denominator and numerator of the current number, we obtain denominator of the next number.

2. If we multiply denominator of the current number by 2 and add to its numerator, we obtain numerator of the next number

```python
start_time = time.time()

count = 0
numerator = 3
denominator = 2

for _ in range(2, 1001):
    numerator, denominator = numerator + 2 * denominator, numerator + denominator
    if len(str(numerator)) > len(str(denominator)):
        count += 1

answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Time Complexity:
   - Loop Iteration: The code contains a loop that iterates from $2$ to $1000$. Therefore, the loop will execute $999$ times. The operations inside the loop, including arithmetic calculations, string conversions, and comparisons, all have constant time complexity.
   - Arithmetic Operations: The arithmetic operations `numerator + 2 * denominator` and `numerator + denominator` have constant time complexity.
   - String Conversions: The string conversions `str(numerator)` and `str(denominator)` have a time complexity proportional to the length of the resulting strings, which is logarithmic in the value of the numbers.
   - Comparison: The comparison `len(str(numerator)) > len(str(denominator))` has a time complexity proportional to the length of the compared strings, which is logarithmic in the value of the numbers.

   Therefore, the overall time complexity of the code is $O(n)$, where $n$ is the number of iterations in the loop (in this case, $999$).

2. Space Complexity:
   - Variables: The code uses a constant number of variables (`count`, `numerator`, `denominator`) that store integer values. The space complexity of these variables is $O(1)$ or constant.
   - String Conversions: The string conversions `str(numerator)` and `str(denominator)` create new string objects, which temporarily increase the space complexity. However, as these strings are not stored or accumulated over time, the overall impact on space complexity is negligible.

   Therefore, the overall space complexity of the code is $O(1)$ or constant.

In summary, the code has a time complexity of $O(n)$ and a space complexity of $O(1)$, where $n$ is the number of loop iterations. The code executes efficiently as it performs a fixed number of operations for each iteration, and the space usage remains constant throughout the execution.
