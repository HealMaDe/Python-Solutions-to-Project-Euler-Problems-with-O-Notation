We need to consider only factors up to the square root of the number. This is because the fact that each factor above the square root has a mirror factor below. We have to consider the perfect square numbers too.

## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
import math   #For calculating square root
```

## Main Solution <a name="Main-Solution"></a>


First we define a function that counts all divisors of any number based on mirror factors.

```python
def count_divisors(n):
    count = 0
    sqrt = int(math.sqrt(n))
    for i in range(1, sqrt +1):
        if n % i == 0:
            count += 2 #because of miror factors
    if sqrt * sqrt == n:
        count -= 1     #because of perfect square
    return count
```

Then we proceed to solve the problem

```python
start_time = time.time()

number = 1
tri_num = 1 #triangular number
while True:
    number += 1
    tri_num += number
    if count_divisors(tri_num)>500:
        break

answer = tri_num

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(\sqrt{n})$**, where $n$ is the input number. This is because the code iterates from $1$ to $\sqrt{n}$  in the *count_divisors* function.

The **space complexity** of my solution is **$O(1)$** because the code only uses a constant amount of additional space to store the count variable and the sqrt variable.
