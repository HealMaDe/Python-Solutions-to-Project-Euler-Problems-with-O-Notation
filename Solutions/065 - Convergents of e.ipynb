## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>

If we look closely, each numerator of the sequence are calculated as follows:

$numerator_{i} = a_i * numerator_{i-1} + numerator_{i-2}$

and if we pay more attention we can see that $a_n$ also follow a pattern:
$1,2*i,1,1,2*i,1,... ; i=1,2,3,...$

```python
start_time = time.time()

numerator = [3,8,11]
i = 2
while i!= 34:
    numerator.append(numerator[-1] + numerator[-2])
    numerator.append(2*i*numerator[-1] + numerator[-2])
    numerator.append(numerator[-1] + numerator[-2])
    i += 1

sum = 0
for i in str(numerator[-1]):
    sum += int(i)

answer = sum

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code consists of a while loop that continues until `i` is equal to 34.
- Inside the while loop, there are constant-time operations such as appending elements to the `numerator` list and incrementing `i`.
- The number of iterations in the while loop is fixed and does not depend on the size of the input.
- Therefore, the **time complexity** of the code is $O(1)$ or constant time.

Space Complexity:
- The space complexity is determined by the space used to store the `numerator` list and a few additional variables.
- The size of the `numerator` list grows with each iteration of the while loop, but it has a fixed upper bound of 34 elements.
- Therefore, the **space complexity** of the code is $O(1)$ or constant space.
