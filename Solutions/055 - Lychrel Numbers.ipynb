## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

count = 0
for i in range(1,10000):
  n = 1
  s1 = str(i)
  s2 = s1[::-1]
  while n <= 50:
    s1 = str(int(s1) + int(s2))
    s2 = s1[::-1]
    if s1 == s2:
      break
    else:
      n += 1
  else:
    count += 1

answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, where $n$ is the range of numbers from 1 to 10000. This is because for each number in the range, the code performs a loop that iterates up to $50$ times. Since the range is $10000$, the worst case scenario is when all numbers in the range require $50$ iterations. Therefore, the time complexity is $O(n^2)$.

The **space complexity** of my solution is **$O(1)$**, as it does not use any additional data structures that grow with the input size. The variables used in the code have constant space requirements.
