## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from sympy import divisors #finding divisors of a number
```

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

n = 28123  #limit\n",
abundants = [i for i in range(n) if sum(divisors(i)[:-1])>i] #generate abundant numbers

#sum of 2 abundant numbers
sum_abundants = {abundants[i] + abundants[j] for i in range(len(abundants)) for j in range(len(abundants))}
answer = sum([i for i in range(n) if i not in sum_abundants])

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Time complexity:

- Generating the list of abundant numbers involves iterating through the numbers up to `n` and calculating the sum of divisors for each number. The `divisors()` function's time complexity depends on its implementation, but a common approach is to iterate up to the square root of the number, making it approximately $O(\sqrt{n})$.

- The list comprehension `[i for i in range(n) if sum(divisors(i)[:-1]) > i]` iterates through the numbers up to `n` and performs the divisors calculation. The time complexity of this part is approximately $O(n * \sqrt{n})$.
        
- The nested loop to calculate the sum of two abundant numbers iterates over the `abundants` list, which has a length of at most `n`. Hence, the time complexity of this part is $O(n^2)$.

- The final step to compute the sum of numbers that are not in `sum_abundants` involves iterating through the numbers up to `n` and checking if each number is present in the `sum_abundants` set. The time complexity of this step is $O(n)$.
        
- Overall, the **time complexity** of the code is approximately $O(n * \sqrt{n} + n^2)$.

<br>

Space complexity:
- The code uses the `abundants` list to store the abundant numbers, which has a length of at most `n`. Therefore, the space complexity of this list is $O(n)$.

- The `sum_abundants` set stores the sums of two abundant numbers. The number of unique sums is at most $O(n^2)$, as there can be duplicates. Hence, the space complexity of this set is $O(n^2)$.

- The space complexity of the code is dominated by the `sum_abundants` set, resulting in $O(n^2)$ space complexity.
