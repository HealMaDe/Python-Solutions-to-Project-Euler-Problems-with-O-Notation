## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>

Simply Define necessary functions

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True


def is_palindrome(n):
    return str(n)==str(n)[::-1]

def is_perfect_square(n):
    root = int(math.sqrt(n))
    return root * root == n
```

Then tackle problem

```python
start_time = time.time()

c =0
sum = 0
n = 1
while c!=50:
  if is_perfect_square(int(str(n**2)[::-1])):
    if is_prime(math.sqrt(n**2)) and is_prime(math.sqrt(int(str(n**2)[::-1]))) and not is_palindrome(n**2):
      sum += n**2
      n += 2
      c += 1
    else:
      n += 2
  else:
    n += 2

answer = sum
end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time complexity:
- The code uses a while loop with a condition `c!=50`, which means it iterates until `c` reaches 50. The loop increments `n` by 2 in each iteration.
- Inside the loop, the code performs several operations such as checking for perfect square, primality, palindrome, and square root calculations.
- The time complexity of checking for perfect square and palindrome is $O(1)$ as they involve simple operations on the number's digits.
- The time complexity of the primality check is $O(\sqrt{n})$ since it iterates up to the square root of `n`.
- The time complexity of the square root calculation using `math.sqrt()` is $O(1)$.
- Overall, the **time complexity** of the code is approximately $O(c * \sqrt{n})$.

Space complexity:
- The code uses a constant amount of extra space for variables `c`, `sum`, and `n`.
- The space complexity is $O(1)$ as it does not use any additional data structures that grow with the input size.

It's important to note that the exact time complexity depends on the specific properties of the numbers being checked and the efficiency of the underlying implementations of functions like `math.sqrt()`. The analysis provided above gives a rough estimation of the code's complexity.
