## Main Solution <a name="Main-Solution"></a>

I solved this in 2 ways.

First I try to sort the list of digits based on the list existed in the file

```python
start_time = time.time()

with open("0079_keylog.txt", 'r') as file:
        login_attempts = [line.strip() for line in file]

L = []
s = set()
for i in login_attempts:
    L.append(i)
    d1, d2, d3 = i
    s.add(d1)
    s.add(d2)
    s.add(d3)

L1 = sorted(L)
L2 = list(s)



sorted_L2 = sorted(L2, key=lambda x: [L1_elem.index(digit) for L1_elem in L1 for digit in str(x) if digit in L1_elem])

answer = "".join(sorted_L2)
end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

my second approach is based on statistics. I sort the digits based on the mean of their appearance in a position.

```python
start_time = time.time()

#statistical approach
with open("0079_keylog.txt", 'r') as file:
        login_attempts = [line.strip() for line in file]

try_code = []
s = set()
for i in login_attempts:
    try_code.append(i)
    d1, d2, d3 = i
    s.add(d1)
    s.add(d2)
    s.add(d3)

digits = list(s)

d = {}
for i in digits:
    d[i]=[]

for code in try_code:
    for index in range(3):
        d[code[index]].append(code.index(code[index]))

for i in d:
    d[i] = statistics.mean(d[i])

passcode = [key for key, _ in sorted(d.items(), key=lambda x: x[1])]

answer = "".join(passcode)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

Let's analyze the time and space complexity of my second approach.

Time Complexity:
1. Iterating over `login_attempts` to populate `try_code` and `s` takes $O(n)$ time, where $n$ is the number of login attempts.
2. Creating the `digits` list from the set `s` takes $O(m)$ time, where $m$ is the number of unique digits across all login attempts.
3. Creating the dictionary `d` with empty lists for each digit takes $O(m)$ time.
4. Iterating over `try_code` and appending indices to the corresponding lists in `d` takes $O(n)$ time.
5. Calculating the mean for each digit in `d` takes $O(m)$ time.
6. Sorting the dictionary `d` based on values takes $O(m * log(m))$ time.
7. Joining the sorted keys in `passcode` takes $O(m)$ time.

Overall, the **time complexity** of the code is $O(n + m + m + n + m* log(m) + m)$, which simplifies to $O(n + m *log(m))$.

Space Complexity:
1. The `try_code` list requires $O(n)$ space to store the login attempts.
2. The `s` set requires $O(m)$ space to store the unique digits.
3. The `digits` list requires $O(m)$ space to store the digits from the set.
4. The dictionary `d` requires $O(m)$ space to store $m$ keys and their corresponding lists.
5. The `passcode` list requires $O(m)$ space to store the sorted keys.

Overall, the **space complexity** of the code is $O(n + m + m + m + m)$, which simplifies to $O(n + 4m)$.
