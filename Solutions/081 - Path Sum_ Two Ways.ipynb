## Main Solution <a name="Main-Solution"></a>

First we build our matrix from the content of the file

```python
matrix = []
with open("0081_matrix.txt", 'r') as file:
    for line in file:
        row = list(map(int, line.strip().split(',')))
        matrix.append(row)
```

Now we tackle the problem

```python
start_time = time.time()

rows = len(matrix)
columns = len(matrix[0])

# Create a memoization table to store the minimum path sums
memory = [[0] * columns for _ in range(rows)]

# Initialize the top-left corner
memory[0][0] = matrix[0][0]

# Initialize the first row
for col in range(1, columns):
    memory[0][col] = memory[0][col - 1] + matrix[0][col]

# Initialize the first column
for row in range(1, rows):
    memory[row][0] = memory[row - 1][0] + matrix[row][0]

# Calculate the minimum path sum for each position
for row in range(1, rows):
    for col in range(1, columns):
        memory[row][col] = min(memory[row - 1][col], memory[row][col - 1]) + matrix[row][col]

# Return the minimal path sum for the bottom-right corner
answer = memory[rows - 1][columns - 1]

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time complexity of my code is $O(rows * columns)$, where rows and columns represent the dimensions of the matrix. This complexity arises from the nested loops used to calculate the minimum path sums for each position in the matrix.

The space complexity of the code is $O(rows * columns)$ as well. This is because the code creates a memoization table (`memory`) of the same size as the matrix to store the minimum path sums. Additionally, the code also creates a matrix (`matrix`) to read the values from the file, which also has a space complexity of $O(rows * columns)$.

Overall, both the **time and space complexity** of the code are determined by the dimensions of the matrix, specifically rows and columns.
