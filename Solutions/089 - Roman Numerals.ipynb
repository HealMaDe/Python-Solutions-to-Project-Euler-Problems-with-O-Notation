## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>


The `re.sub(pat, replacement, str)` function searches for all the instances of pattern in the given string, and replaces them

```python
start_time = time.time()

File = open('roman.txt','r')

answer = 0
for roman in File:
	cache = len(roman)

	roman = re.sub('IIII','IV',roman)
	roman = re.sub('XXXX','XL',roman)
	roman = re.sub('CCCC','CD',roman)
	roman = re.sub('VIV','IX',roman)
	roman = re.sub('LXL','XC',roman)
	roman = re.sub('DCD','CM',roman)

	answer += cache - len(roman)


end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- Reading the file: The time complexity of reading a file is typically considered to be $O(n)$, where $n$ is the size of the file. In this case, the code reads the file line by line, so the time complexity would be proportional to the number of lines in the file.
- Regular expression substitutions: The regular expression substitutions in the code use the `re.sub` function. The time complexity of this function depends on the size of the input string and the complexity of the regular expression pattern. In general, the time complexity of regular expression matching is considered to be $O(m*n)$, where $m$ is the length of the pattern and $n$ is the length of the input string. Since the provided code performs several substitutions in a loop, the overall time complexity of the substitutions would be proportional to the total length of the input strings in the file.

Overall, the **time complexity** of the code can be approximated as $O(n * m * k)$, where $n$ is the number of lines in the file, $m$ is the average length of each line, and $k$ is the number of substitutions performed.

Space Complexity:

The **space complexity** of the code mainly depends on the size of the input file and the regular expression substitutions. However, the space complexity of the code is relatively low because it does not use any significant additional data structures to store intermediate results. The space complexity can be considered as $O(1)$ or constant space.
