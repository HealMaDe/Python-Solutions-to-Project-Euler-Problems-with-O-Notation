## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

cycle = 0 #cycle of repeating
max = 0 #maximum cycle

for d in range(2, 1000):
  R = [0] * d #list of residuals
  numerator = 1
  position = 0

  while R[numerator] == 0 and numerator != 0:
    R[numerator] = position
    numerator = numerator * 10 % d
    position += 1

    if position - R[numerator] > cycle:
      cycle = position - R[numerator]
      max = d

answer = max

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time complexity:

- The code uses a loop that iterates through the numbers from 2 to 999 (range 2, 1000). This loop has $O(1)$ time complexity since the range is fixed.

- Inside the loop, there is another loop that iterates until a repeating pattern is found. The number of iterations in this loop depends on the number `d` being processed.

- The inner loop iterates until either a repeating pattern is found (condition: `R[numerator] == 0`) or the numerator becomes zero (condition: `numerator != 0`).

- Since the value of `numerator` is reduced by a factor of 10 in each iteration, the number of iterations in the inner loop is at most `d` (the value of `d` being processed).

- Therefore, the worst-case time complexity of the inner loop is $O(d)$.

- Overall, the **time complexity** of the code is approximately $O(1) + O(d) * O(1)$ (nested loop).

<br>

Space complexity:

- The code uses two integer variables (`cycle` and `max`), which require constant space.

- It also uses a list `R` of size `d` to store the residuals. The space complexity of this list is $O(d)$.

- The **space complexity** of the code is dominated by the `R` list, resulting in $O(d)$ space complexity.
