## Main Solution <a name="Main-Solution"></a>
First we take care of the prime part of the problem with the following definition:

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5)+ 1):
        if n%i == 0:
            return False
    return True
```

Then we try to define a function that check the circularity of any given number:

```python
def circular_prime(n):
    L = list(str(n))
    up = len(L)
    for i in range(up):
        s = "".join(L)
        if not is_prime(int(s)):
                return False

        L.append(L.pop(0)) #build circular list

    return True
```

Now we can solve the problem by counting circular Prime numbers.

```python
start_time = time.time()

limit = 10**6
count = 0
for i in range(1,limit,2): #skip even numbers ..... remember number 2
    if str(i)[0] in "24685":  #remember number 5
        continue
    elif circular_prime(i):
        count += 1

answer = count + 2 #numbers 2 and 5 are missing in our loop

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n^2)$**, because for each number from $1$ to the $limit$, we check if it is a circular prime by calling the is_prime function and rotating the digits. The is_prime function has a time complexity of $O(\sqrt{n})$ because it checks if the number is divisible by any number up to its square root. Since we call the is_prime function for each rotation of the digits, the overall time complexity is $O(n^2)$.


The **space complexity** of my solution is **$O(log(n))$** because we store the digits of the number in a list. The number of digits in a number is given by $log(n)$, so the space complexity is logarithmic.
