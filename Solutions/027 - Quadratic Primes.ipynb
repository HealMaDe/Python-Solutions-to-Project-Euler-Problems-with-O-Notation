## Main Solution <a name="Main-Solution"></a>
Simply Define necessary functions

```python
def isprime(n):
    if n < 2:
        return False
    for i in range(2,int(n**0.5) + 1):
        if n%i==0:
            return False
    return True


#count_consecutive_primes
def count_primes(a, b):
    n = 0
    while isprime(n**2 + a*n + b):
        n += 1
    return n
```

Then tackle the problem

```python
start_time = time.time()

max = 0
answer = 0

for a in range(-999, 1000):
    for b in range(-1000, 1001):
        primes = count_primes(a, b)
        if primes > max:
            max = primes
            answer = a * b

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time complexity:
- The `isprime()` function checks if a number `n` is prime. It iterates from 2 up to the square root of `n` and performs a divisibility check for each iteration. Therefore, the time complexity of this function is approximately $O(\sqrt{n})$.
- The `count_primes()` function calls `isprime()` in a loop until a non-prime value is encountered. The number of iterations in this loop depends on the value of `n`, which increases until a non-prime value is found. The worst-case time complexity of this loop is $O(n)$, where `n` is the number of consecutive primes found.
- The nested loops iterate over the values of `a` and `b` within a certain range. The outer loop has a fixed range of -999 to 999 (1999 iterations), and the inner loop has a fixed range of -1000 to 1000 (2001 iterations). Therefore, the overall time complexity of the nested loops is $O(1999 * 2001)$.

- Overall, the **time complexity** of the code is approximately $O(\sqrt{n} * n * 1999 * 2001)$.

Space complexity:
- The code uses a few integer variables (`n`, `max`, and `answer`), which require constant space.
- There are no significant additional data structures used that grow with the input size.

- The **space complexity** of the code is $O(1)$ as it uses a constant amount of extra space.
