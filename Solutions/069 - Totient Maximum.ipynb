## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
import sympy  #for next prime
```

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

prime = 2
answer = 2
while True:
    answer = answer*sympy.nextprime(prime)
    prime = sympy.nextprime(prime)
    if answer >1000000:
      answer //= prime
      break

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
- The code consists of a `while` loop that continues indefinitely until a certain condition is met.
- Inside the loop, the code calls `sympy.nextprime()` twice to find the next prime number and calculate the updated value of `answer`.
- The `sympy.nextprime()` function is based on the prime number generation algorithms, such as the Sieve of Eratosthenes.
- The time complexity of generating the next prime number using these algorithms is typically sublinear.
- Therefore, the time complexity of each iteration of the `while` loop is sublinear or $O(log(n))$, where $n$ is the current value of `prime`.

- The loop continues until `answer` exceeds $1,000,000$, and at that point, it performs a division and breaks the loop.
- The number of iterations required to reach this condition depends on the prime numbers and their products, but it is difficult to determine the exact number of iterations without specific input values.

- Overall, the **time complexity** of the code can be considered as approximately $O(log(n))$, where $n$ is the value of the largest prime number encountered during the execution.

Space Complexity:
- The code uses a constant amount of space for variables `prime` and `answer`.
- The `sympy.nextprime()` function does not require any significant additional space as it generates primes on the fly.

- Therefore, the space complexity of the code is $O(1)$ since it uses a fixed amount of space regardless of the input size.
