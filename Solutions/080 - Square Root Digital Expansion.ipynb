## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from decimal import *  #working with decimals
```

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

getcontext().prec = 102 #set the precision to 102 decimal places
s = ""
sum = 0
for num in range(2,100):
    #check not to be aperfect square
    if int(num ** 0.5) ** 2 != num:
        #collecting all digits aith deletion of decimal point
        s += str(Decimal(num).sqrt()).replace(".","")[:100]

sum = 0
for i in s:
    sum += int(i)

answer = sum

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the range of numbers from $2$ to $100$. This is because the code iterates through each number in the range and performs a constant number of operations for each number.

The **space complexity** of my solution is **$O(1)$**, as it only uses a constant amount of space to store the sum variable and the temporary string $s$. The size of these variables does not depend on the input size.
