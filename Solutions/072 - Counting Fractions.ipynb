## Main Solution <a name="Main-Solution"></a>

All we need to do is to calculate the sum of Totients from $2$ to $10^6$. So first we define our totient function.

```python
def totient(n):
    phi = list(range(n + 1))
    is_prime = [True] * (n + 1)

    for p in range(2, n + 1):
        if is_prime[p]:
            for i in range(p, n + 1, p):
                phi[i] -= phi[i] // p
                is_prime[i] = False

    return phi
```

Now we tackle the problem

```python
start_time = time.time()

phi = totient(10**6)
count = sum(phi[2:])  # Exclude phi[0] and phi[1]
answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

Time Complexity:
1. `totient()` function: The function calculates the Euler's totient function values for numbers up to `n` using the sieve algorithm. The time complexity of this function is $O(n *log (log (n)))$, as it performs operations for each prime number up to `n`.

2. `sum()` function: The function iterates over the `phi` list, which has a length of `n+1`. Therefore, the time complexity of this function is $O(n)$.

Therefore, the overall **time complexity** of the code is dominated by the `totient()` function, which is $O(n *log (log (n)))$.

Space Complexity:
1. `totient()` function: The function uses two lists, `phi` and `is_prime`, each with a length of `n+1`. Therefore, the space complexity of this function is $O(n)$ for the `phi` list and $O(n)$ for the `is_prime` list.

2. Other variables: The code uses additional variables like `count` and `answer`, which store integer values. These variables require a constant amount of space, so the space complexity of these variables is $O(1)$.

Therefore, the overall **space complexity** of the code is dominated by the `totient()` function, which is $O(n)$ in terms of the upper limit `n`.
