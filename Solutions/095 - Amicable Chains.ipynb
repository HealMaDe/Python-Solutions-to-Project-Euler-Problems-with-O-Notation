$**Hint**$

Write a function based on sieve Algorithm for generating sum of proper divisors

## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
```

## Main Solution <a name="Main-Solution"></a>

First we define our divisor generator function based on sieve.

```python
def generate_sum_div(limit):

    sieve = [0]*limit
    for i in range(1,limit):
        for j in range(2*i, limit, i):
            sieve[j] += i

    return sieve
```

Now we tackle the problem

```python
start_time = time.time()

divisors = generate_sum_div(10**6)

memory = {}

def count_term(n):
    t = n
    s = set()
    while t not in s:
        s.add(t)
        t = divisors[t]

        if t >10**6 or t == 1: #skip prime numbers and exceed condition
            memory[n] = (0,0)
            return memory[n]

    if t == n:
        memory[n] = (len(s),min(s))
        return memory[n]
    else:
        memory[n] = (0,0)
        return memory[n]

maximum = 0
for i in range(1,10**6):
    if count_term(i)[0] > maximum:
        maximum = count_term(i)[0]
        answer = i

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

The `generate_sum_div` function:
- Time complexity: $O(n *log (n))$
  - The outer loop runs for `n` iterations.
  - The inner loop runs for approximately `log(n)` iterations for each value of `i`.
  - Therefore, the overall time complexity is approximately $O(n *log (n))$.

  Space complexity: $O(n)$
  - The `sieve` list has a length of `limit`, so it requires $O(n)$ space.

The `count_term` function:
- Time complexity: $O(k)$ (where $k$ is the length of the sequence until a repetitive term is found)
  - In the worst case, the `while` loop iterates until a repetitive term is found, which can take up to `k` iterations.
  - Each iteration performs constant time operations.
  - Therefore, the time complexity is $O(k)$.

- Space complexity: $O(k)$
  - The `s` set stores the visited terms until a repetitive term is found, which can take up to `k` elements.
  - Therefore, the space complexity is $O(k)$.

The main loop:
- Time complexity: $O(n)$
  - The loop iterates `n` times, where `n` is the upper limit of the range.
  - The `count_term` function is called for each iteration, which has a time complexity of $O(k)$ as explained above.
  - Therefore, the overall time complexity is $O(n * k)$.

- Space complexity: $O(n)$
  - The `memory` dictionary stores the computed results for each term up to `n`, which can take up to `n` elements.
  - Therefore, the space complexity is $O(n)$.

In summary:

- The **time complexity** of the code is dominated by the main loop and is approximately $O(n * k)$, where `n` is the upper limit of the range and `k` is the length of the sequence until a repetitive term is found.
- The **space complexity** of the code is approximately $O(n)$ due to the `memory` dictionary and the `sieve` list.
