## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

answer = 0
for i in range(10**6):
    if str(i) == str(i)[::-1] and bin(i)[2:] == bin(i)[2:][::-1]:
        answer += i

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the range of the for loop ($10^6$). This is because the code iterates through the range of numbers from $0$ to $10^6$ and performs constant time operations for each iteration.

The **space complexity** of my solution is **$O(1)$**, as it only uses a constant amount of space to store the variable $answer$ and the loop variable $i$. The space used does not depend on the size of the input.
