## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
from itertools import permutations  #generate permutations
```

## Main Solution <a name="Main-Solution"></a>

```python
start_time = time.time()

numbers = list(range(10))  # Numbers from 0 to 9

summ = 0
L = [2,3,5,7,11,13,17]

# Iterate over permutations and filter out those with 0 in the first position
for i in permutations(numbers):
    count = 0
    if i[0] != 0:
        s = ''.join(str(j) for j in i)

        for x in range(7):
            if int(s[x+1:x+4])%L[x] == 0:
                count += 1

            else:
                break
        if count == 7:
            summ += int(s)

answer = summ

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n!)$**, because it iterates over all permutations of the numbers from $0$ to $9$. The number of permutations is factorial of $n$, where $n$ is the number of elements in the list.

The **space complexity** of my solution is **$O(n)$**, as it creates a list of numbers from $0$ to $9$, which has a size of $n$. Additionally, it creates a string 's' to store each permutation, which also has a size of $n$.
