## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

# Set of numbers from 0 to 9
available_numbers = set(range(10))

# Initialize count variable
count = 0

# Digits for cube1 to form square numbers
cube1_digits = [0, 0, 0, 1, 2, 3, 6, 8]

# Digits for cube2 to form square numbers
cube2_digits = [1, 4, 9, 6, 5, 6, 4, 1]

# Nested loops to generate combinations for cube1
for digit1 in range(10):
    for digit2 in range(digit1 + 1, 10):
        for digit3 in range(digit2 + 1, 10):
            for digit4 in range(digit3 + 1, 10):
                cube1_remaining = available_numbers.difference([digit1, digit2, digit3, digit4])

                # Nested loops to generate combinations for cube2
                for digit5 in range(10):
                    for digit6 in range(digit5 + 1, 10):
                        for digit7 in range(digit6 + 1, 10):
                            for digit8 in range(digit7 + 1, 10):
                                cube2_remaining = available_numbers.difference([digit5, digit6, digit7, digit8])

                                # Check if 6 or 9 is present in combinations for cube1 or cube2 and update the sets accordingly
                                if 9 in cube1_remaining or 6 in cube1_remaining:
                                    cube1_remaining.update([6, 9])
                                if 9 in cube2_remaining or 6 in cube2_remaining:
                                    cube2_remaining.update([6, 9])

                                satisfies_conditions = True  # Flag to track if the combination satisfies the conditions

                                # Iterate through cube1_digits and cube2_digits simultaneously and check if the corresponding digits are present in the sets
                                for digit_cube1, digit_cube2 in zip(cube1_digits, cube2_digits):
                                    satisfies_conditions &= (digit_cube1 in cube1_remaining and digit_cube2 in cube2_remaining) or \
                                                            (digit_cube1 in cube2_remaining and digit_cube2 in cube1_remaining)

                                # If the combination satisfies the conditions, increment the count
                                if satisfies_conditions:
                                    count += 1

answer = count//2  #Divide the count by 2 to account for counting each pair twice

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my code is $O(n^8)$, where $n$ is the range of numbers (10 in this case). This is because there are eight nested loops, each iterating from $0$ to $n-1$. However, since $n$ is a constant value (10), we can consider the time complexity to be constant, i.e., $O(1)$.

The **space complexity** of the code is also constant, $O(1)$. This is because the code uses a fixed amount of space to store the set of available numbers, cube digits, and some boolean flags. The space required does not depend on the input size or any iterative variables.
