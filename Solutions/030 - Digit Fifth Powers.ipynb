## Required Modules <a name="Required-Modules"></a>

## Main Solution <a name="Main-Solution"></a>



```python
start_time = time.time()

answer = 0
for i in range(10,195000):
    s = str(i)
    x = 0
    for j in range(0,len(s)):
        x += int(s[j])**5
    if x == i:
        answer += i


end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n*m)$**, where $n$ is the range from $10$ to $195000$ and $m$ is the number of digits in each number. This is because we have two nested loops - the outer loop iterates over the range from $10$ to $195000$, and the inner loop iterates over the digits of each number. The number of digits in each number is determined by the length of the string representation of the number.

The **space complexity** of my solution is **$O(1)$** because it only uses a constant amount of space to store the variables $answer$, $i$, $s$, and $x$. The space used does not depend on the input size.
