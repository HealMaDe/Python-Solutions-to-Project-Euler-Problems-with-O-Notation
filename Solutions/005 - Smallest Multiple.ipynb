**$Hint$**

If the number is supposed to be divisible by all the numbers from $1$ to $20$, it is also has to be divisible by multiple of all primes between $1$ to $20$ $(2*3*5*7*11*13*17*19)$.

## Required Modules <a name="Required-Modules"></a>

```python
import time   #Capturing the execution time
```

## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

n = 2*3*5*7*11*13*17*19 #initial guess
L = []
while True:
    for i in range(1,21):
        L.append(n%i)
    if sum(L) == 0:
        break
    else:
        del L
        L = []
        n += 2*3*5*7*11*13*17*19
answer = n

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The **time complexity** of my solution is **$O(n)$**, where $n$ is the value of the final number that satisfies the condition. This is because the code iterates through each number from $1$ to $20$ and performs a constant amount of operations for each number.

The **space complexity** of my solution is **$O(1)$**, as it only uses a fixed amount of memory to store the variables $n$ and $L$. The size of the input does not affect the amount of memory used by the code.
