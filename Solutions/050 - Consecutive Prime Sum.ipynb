## Main Solution <a name="Main-Solution"></a>
First we generate all primes based on [Sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes) Algorithm.

```python
def generate_primes_below(n):
    sieve = [True]*(n+1)
    sieve[0] = sieve[1] = False  #0 and 1 are not primes
    p = 2
    while p**2<n:
        for i in range(p**2,n+1,p):
            sieve[i] = False

        p += 1
    primes = [p for p in range(len(sieve)) if sieve[p]]

    return primes
```

Then we tackle the main problem

```python
start_time = time.time()

limit = 10**6
primes = generate_primes_below(limit)

longest_seq = []
seq_len = 0

for i in range(len(primes)):
    for j in range(i + seq_len, len(primes)):
        seq = primes[i:j+1]

        if sum(seq) > limit:
            break

        if sum(seq) in primes:
            longest_seq = seq
            seq_len = len(seq)

answer = sum(longest_seq)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png


**Time Complexity**:

1. **Generating Primes**: The function uses the Sieve of Eratosthenes algorithm to generate prime numbers up to $n$. The time complexity of this function is approximately $O(n*log(log(n)))$ as it involves marking composite numbers in the sieve. This step has a significant impact on the overall time complexity.

2. **Nested Loops**: The outer loop iterates $len(primes)$ times, which is approximately $O(n/log(n))$. The inner loop iterates over a range that depends on $i$ and seq_len, which can also be up to $O(n/log(n))$ in the worst case. Inside the loops, calculating the sum of the sequence and checking if it is in the primes list both have a constant time complexity.

Therefore, the overall time complexity of my code is approximately $O(n^2/log(n))$.


<br>

**Space Complexity**:

1. **Sieve**: The function creates a sieve list of size $n+1$. Hence, the space complexity of the sieve is $O(n)$.

2. **primes List**: The function also creates a primes list to store the generated prime numbers. The size of this list depends on the number of prime numbers up to $n$, which is approximately $n/log(n)$. Therefore, the space complexity of the primes list is $O(n/log(n))$.

3. **Longest Sequence**: The longest_seq list stores the longest sequence of consecutive primes found. The size of this list depends on the length of the sequence, which can be up to $O(n/log(n))$.


Overall, the space complexity of my code is $O(n)$ due to the sieve and primes list. The space complexity of the longest_seq list is dominated by the length of the sequence, which is relatively small compared to $n$ in this case.
