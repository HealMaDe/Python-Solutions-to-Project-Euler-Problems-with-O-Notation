## Main Solution <a name="Main-Solution"></a>


```python
start_time = time.time()

def find_least_number():
    bouncy_count = 0
    total_numbers = 99  # Start from 100
    proportion = 0

    while proportion != 0.99:
        total_numbers += 1

        if is_bouncy(total_numbers):
            bouncy_count += 1

        proportion = bouncy_count / total_numbers

    return total_numbers


def is_bouncy(number):
    digits = list(str(number))

    increasing = False
    decreasing = False

    for i in range(1, len(digits)):
        if digits[i] > digits[i - 1]:
            increasing = True
        elif digits[i] < digits[i - 1]:
            decreasing = True

        if increasing and decreasing:
            return True

    return False


# Find the least number for which the proportion of bouncy numbers is exactly 99%
answer = find_least_number()

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

1. `find_least_number` function: The time complexity of this function depends on the number of iterations required to find the least number for which the proportion of bouncy numbers is exactly 99%. In the worst case, this could potentially require iterating through a large number of numbers until the condition is met. However, since there is no specific upper bound mentioned in the code, it is difficult to provide an exact time complexity. We can consider the time complexity as $O(N)$, where $N$ represents the number of iterations until the condition is met.

2. `is_bouncy` function: The time complexity of this function depends on the number of digits in the input number. It iterates through the digits once, comparing each digit with the previous one. Therefore, the time complexity can be considered as $O(d)$, where $d$ represents the number of digits in the input number.

Overall, the **time complexity** of the code is primarily determined by the `find_least_number` function, which can be approximated as $O(N)$, where $N$ represents the number of iterations until the condition is met.
