## Main Solution <a name="Main-Solution"></a>
If you consider the problem carefully, you must make a total of 20 right movements and 20 down movements, regardless of the order. The task is to find all the possible arrangements of 20 rights (R) and 20 downs (D). If we treat these movements as elements, we have a total of 40 elements, with 20 of them being R and the other 20 being D. Let's explore the different combinations.

Imagine you have one particular combination. By swapping any of the D's with each other, you will still end up with the same combination. The same applies to the R's.

So the solution is 40 choose 20

```python
start_time = time.time()

answer = comb(40,20)

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.
