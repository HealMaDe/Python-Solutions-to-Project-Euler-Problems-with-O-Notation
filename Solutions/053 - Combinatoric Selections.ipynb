## Main Solution <a name="Main-Solution"></a>
First we define our two functions

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n*fact(n-1)

def c(n,r):
    return fact(n)/(fact(r)*fact(n-r))
```

Then we tackle the main problem

```python
start_time = time.time()

count = 0
for n in range(1,101):
    for r in range(1,101):
        if r>n:
            continue
        else:
            if c(n,r)>1000000:
                count += 1
answer = count

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Time Complexity:
   - Factorial Function (`fact`): The `fact` function calculates the factorial of a given number `n` recursively. The time complexity of this function is $O(n)$ since it needs to perform `n` multiplications to calculate the factorial.
   - Combination Function (`c`): The `c` function calculates the combination (nCr) using the factorial function. Since it involves calling the factorial function three times, the time complexity of this function is also $O(n)$.
   - Nested Loops: The code contains nested loops that iterate through the range from $1$ to $100$ for both `n` and `r`. Therefore, the nested loops have a time complexity of $O(n^2)$.

   Overall, the time complexity of the code is $O(n^3)$ due to the nested loops and the factorial calculations inside the combination function.

2. Space Complexity:
   - Recursive Calls: The `fact` function uses recursion to calculate the factorial. Each recursive call adds a new frame to the call stack, consuming additional space. Since the maximum recursion depth is limited by the input value of `n` (up to $100$ in this case), the space complexity of the recursive calls is $O(n)$.
   - Additional Variables: The code uses additional variables to store the count and loop variables. These variables have a constant space complexity of $O(1)$.

   Therefore, the overall space complexity of the code is $O(n)$ due to the recursive calls.
