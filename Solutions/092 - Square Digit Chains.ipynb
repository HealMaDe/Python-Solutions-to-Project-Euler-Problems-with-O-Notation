## Main Solution <a name="Main-Solution"></a>

Let's difine a function to calculate square digits of any given number. Remember first we have to define a memory to use it at a look up table.  

```python
memory = {"1":set(),
          "89":set()}

def square_digit(n):
    while n != 1 and n != 89:
        n = sum(int(digit)**2 for digit in str(n))
        if n in memory["1"]:
            return "1"
        elif n in memory["89"]:
            return "89"
    return str(n)
```

Now we try to solve the problem

```python
start_time = time.time()

n = 1
count = 0
while n!= 10**7:
    t = square_digit(n)
    memory[t].add(n)
    n += 1

answer = len(memory["89"])

end_time = time.time()
execution_time = end_time - start_time

print("Final Answer:",answer,"\nExecution time:","{:.4f}".format(execution_time), "seconds")
```

## Big O Notation <a name="Big-O-Notation"></a>
Big O Notation is a tool used to describe the **time complexity of algorithms**. It calculates the time taken to run an algorithm as the input grows. In other words, it calculates **the worst-case time complexity of an algorithm**.

2023-11-09_162722.png

The time and space complexity of my code can be analyzed as follows:

1. Function `square_digit`:
   - The function performs a constant number of operations on `n`, including converting it to a string, iterating over its digits, and calculating the sum of squared digits.
   - The time complexity of the function is determined by the number of digits in `n`, denoted as `d`. In the worst case, `d` can be $O(log(n))$, where $n$ is the value of `n`.
   - Therefore, the time complexity of the function is $O(log(n))$.

2. Building the `memory` dictionary:
   - The code initializes the `memory` dictionary with two empty sets.
   - The time complexity of this operation is $O(1)$ since it involves a constant number of operations.

3. Looping from `n = 1` to `n = 10**7`:
   - The code iterates over a range of numbers from $1$ to $10^7$.
   - Within the loop, it calls the `square_digit` function and performs set operations on the `memory` dictionary.
   - The overall time complexity of this loop is $O(n * log (n))$ since calling `square_digit` has a time complexity of $O(log (n))$, and the loop iterates $n$ times.

4. Calculating the `answer`:
   - The code retrieves the length of the set `memory["89"]`, which has a constant number of elements.
   - The time complexity of this operation is $O(1)$ since it involves accessing the length of a set.

In summary, the overall **time complexity** of the code is $O(n * log (n))$, where n is the value of `n` in the loop. The dominant factor is the loop that iterates from 1 to 10^7 and calls the `square_digit` function.

Regarding space complexity, the code uses additional data structures such as the `memory` dictionary and the sets within it. The size of these data structures depends on the number of elements stored and can grow up to $10^7$ elements in the worst case. Therefore, the **space complexity** of the code is $O(n)$, where n is the value of `n` in the loop.
